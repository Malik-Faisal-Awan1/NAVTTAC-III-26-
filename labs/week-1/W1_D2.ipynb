```python
per = float(input("Enter your percentage"))
per = round(per)

print(per)

if per >= 86 and per <= 100:
    grade = "A"
elif per >= 82:
    grade = "A-"
elif per >= 78:
    grade = "B+"
elif per >= 74:
    grade = "B-"
elif per >= 70:
    grade = "C+"
elif per >= 66:
    grade = "C"
elif per >= 62:
    grade = "C-"
elif per >= 58:
    grade = "D+"
elif per >= 54:
    grade = "D"
else:
    grade = "F"

print(grade)
```
