```python
nums = [2, 5, 8, 11, 14, 17]
evens = [n for n in nums if n % 2 == 0]
print(evens)  # [2, 8, 14]

nums = [2, 5, 8, 11, 14, 17]
evens = [int(n/2) for n in nums if n % 2 == 0]
print(evens)  # [2, 8, 14]
```

```python
nums= [1,2,3,4,5,6,7,8,9]

odd_cube = [n**3 for n in nums if n%2 != 1]

print (odd_cube)
```

```python
nums= [1,2,3,4,5,6,7,8,9]

even_odd = [n**2 if n%2 == 1 else n**3 for n in nums]

print (even_odd)
```

```python
nums= [1,2,3,4,5,6,7,8,9]

a = {n:n**2 for n in nums if n%2 == 0}

print (a)
```

```python
marks = {"Maria": 92, "Ali": 78, "Zain": 88, "Sara": 95}
top = {name: 'A' if m >= 90 else 'B' for name, m in marks.items()}
print(top)  # {'Maria': 92, 'Sara': 95}
```
