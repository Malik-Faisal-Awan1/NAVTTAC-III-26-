```python
def cash_withdrawl(bal):
    c_wi = input("Enter Amount to be withdrawn: ")
    if(c_wi.isdigit()):
        c_wi = int(c_wi)
        if(bal >= c_wi):
            bal -= c_wi
            print("Cash Withdrawn Successfully! ")
            print("Remaining Balance: ", bal)
        else:
            print("Invalid Input | Try again")
    else:
         print("Invalid Input | Try again")

def cash_deposit(bal):
    c_de = input("Enter Amount to be Deposited: ")
    if(c_de.isdigit()):
        c_de = int(c_de)
        if(c_de >= 0):
            bal += c_de
            print("Cash Deposited Successfully! ")
            print("Available Balance: ", bal)
        else:
            print("Invalid Input | Try again")
    else: 
        print("Invalid Input | Try again")
        
def balance_check(bal):
    print("Your Available Balance is: ", bal)
    
b_Bal = 10000
while(True):
    print("Welcome To Faysal Bank")
    print("1. Cash Withdrawal")
    print("2. Cash Deposit")
    print("3. Cash Check Balance")
    print("4. Exit")
    check = input("Enter Number: ")
    if(check == '1'):
        cash_withdrawl(b_Bal)
    elif(check == '2'):
        cash_deposit(b_Bal)
    elif(check == '3'):
        balance_check(b_Bal)
    else:
        break
```
