```python
def unique_subjects(subjects_line):
    parts = subjects_line.split(", ")
    result = []
    seen = set()
    for s in parts:
        s = s.strip().lower()
        if s not in seen:
            seen.add(s)
            result.append(s)
    return result

subjects_line = ("Python, ai, python, Web, ai")
output = unique_subjects(subjects_line)
print("Unique subjects_line:",output)
```

```python
def clean_name(name):
    return name.strip().title()


def summarize_records(records):
    summary = {}
    
    for name, subjects, marks in records:
        clean = clean_name(name)
        
        seen = set()
        unique = []
        for s in subjects:
            s = s.lower()
            if s not in seen:
                seen.add(s)
                unique.append(s)
        
        avg = round(sum(marks) / len(marks), 2)
        best = max(marks)
        result = "Pass" if avg >= 60 else "Fail"
        
        summary[clean] = {
            "subjects": unique,
            "average": avg,
            "best_mark": best,
            "result": result
        }
    
    return summary


records = [
    ("  ali ", ("python", "web", "python"), [78, 82, 91]),
    ("maria", ("ai", "python"), [92, 88]),
    (" Zain", ("web", "ai", "python"), [60, 73, 69]),
    ("sara  ", ("web",), [49])
]

summary = summarize_records(records)

print("Summary of Records")
print("------------------")
for k, v in summary.items():
    print(k, ":", v)
```

```python
def clean_name(name):
    return name.strip().title()


def build_report(name, *marks, bonus=0):
    clean = clean_name(name)
    
    final_marks = []
    for m in marks:
        m = m + bonus
        if m > 100:
            m = 100
        final_marks.append(m)
    
    avg = round(sum(final_marks) / len(final_marks), 2)
    highest = max(final_marks)
    status = "Pass" if avg >= 50 else "Fail"
    
    return {
        "name": clean,
        "final_marks": final_marks,
        "average": avg,
        "highest": highest,
        "status": status
    }


name = "    uMer    aLi"
marks = 72, 85, 94
bonus = 5

report = build_report(name, *marks, bonus=bonus)

print("Single Student Report")
print("---------------------")
print("Name:", report["name"])
print("Final Marks:", report["final_marks"])
print("Average:", report["average"])
print("Highest:", report["highest"])
print("Status:", report["status"])
```
