```python
def min_max(nums):
    return min(nums), max(nums)

nums = [1,2,3,4,5,6,7,8]
minimum, maximum = min_max(nums)
print(minimum, maximum)
```

```python
def is_strong_password(pw):
    num = False
    up = False
    lo = False
    
    if(len(pw)<8):
       print("Weak Password | Length is not >= 8")
    
    for i in pw:
        if i.isdigit():
            num = True
        elif i.isupper():
            up = True
        elif i.islower():
            lo = True

    if num != True:
        print("Week Password | No Number found!")
    if up != True:
        print("Week Password | No Capital character found!")
    if lo != True:
        print("Week Password | No Lower character found!")
    if num and lo and up: 
        print("Strong Password | saved sussefully!")


pw = input("Enter your Password: ")
ans = is_strong_password(pw)
```
