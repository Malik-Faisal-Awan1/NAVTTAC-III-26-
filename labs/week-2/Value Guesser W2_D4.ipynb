```python
import random
r_val = random.randint(1, 100)
count = 1
while(True):
    val = input("Guess the value: ")
    if(val.isdigit()):
        val = int(val)
        if(val == r_val):
            High_Score = count
            print("Number is equal | You Won! | High Score: ", High_Score)
            break
        elif(val < r_val):
            print("Number is Higher")
            count += 1
        elif(val > r_val):
            print("Number is Lower")
            count += 1
    else:
        print("Wrong Input")
```
