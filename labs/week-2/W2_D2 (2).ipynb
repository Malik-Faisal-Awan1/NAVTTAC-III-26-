```python
pas = input("Enter your Password: ")

num = False
up = False
lo = False

if(len(pas)<8):
   print("Weak Password | Length is not >= 8")

for i in pas:
    if i.isdigit():
        num = True
    elif i.isupper():
        up = True
    elif i.islower():
        lo = True
        
if num != True:
    print("Week Password | No Number found!")
if up != True:
    print("Week Password | No Capital character found!")
if lo != True:
    print("Week Password | No Lower character found!")
if num and lo and up: 
    print("Strong Password | saved sussefully!")
```
