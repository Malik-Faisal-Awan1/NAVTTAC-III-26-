```python
b_Bal = 10000
while(True):
    print("Welcome To Faysal Bank")
    print("1. Cash Withdrawal")
    print("2. Cash Deposit")
    print("3. Cash Check Balance")
    print("4. Exit")
    check = input("Enter Number: ")
    if(check == '1'):
        c_wi = input("Enter Amount to be withdrawn: ")
        if(c_wi.isdigit()):
            c_wi = int(c_wi)
            if(b_Bal >= c_wi):
                b_Bal -= c_wi
                print("Cash Withdrawn Successfully! ")
                print("Remaining Balance: ", b_Bal)
            else:
                print("Invalid Input | Try again")
        else:
            print("Invalid Input | Try again")
    elif(check == '2'):
        c_de = input("Enter Amount to be Deposited: ")
        if(c_de.isdigit()):
            c_de = int(c_de)
            if(c_de >= 0):
                b_Bal += c_de
                print("Cash Deposited Successfully! ")
                print("Available Balance: ", b_Bal)
            else:
                print("Invalid Input | Try again")
        else: 
            print("Invalid Input | Try again")
    elif(check == '3'):
        print("Your Available Balance is: ", b_Bal)
    else:
        break
```
