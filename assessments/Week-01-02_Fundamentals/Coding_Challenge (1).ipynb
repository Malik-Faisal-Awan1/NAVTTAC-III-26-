## Challenge 1 — String Cleaning Pipeline (5 marks)

Write a function `clean_text(raw)` that:
1. Strips leading/trailing whitespace
2. Converts to lowercase using `casefold()`
3. Replaces multiple internal spaces with a single space
4. Returns the cleaned string

**Example:** `clean_text('  Hello   World  ')` → `'hello world'`

```python
# ─── YOUR CODE HERE ───────────────────────────────────────────
def clean_text(raw: str) -> str:
    cleanRaw= raw.casefold().strip().split()
    cleanRaw= (" ").join(cleanRaw)
    return cleanRaw
# ──────────────────────────────────────────────────────────────
```

```python
# --- CHECK ---
test_cases = [
    ('  Hello   World  ', 'hello world'),
    ('  PYTHON   ', 'python'),
    ('already clean', 'already clean'),
    ('  Multiple   Spaces   Here  ', 'multiple spaces here'),
]
passed = 0
for inp, expected in test_cases:
    result = clean_text(inp)
    status = '✅' if result == expected else f'❌ got {repr(result)}'
    print(f'{status}  clean_text({repr(inp)}) == {repr(expected)}')
    if result == expected: passed += 1
print(f'\nScore: {passed}/{len(test_cases)} — {passed * 1.25:.1f}/5 marks')
```

## Challenge 2 — Word Frequency Counter (5 marks)

Write a function `word_freq(text)` that:
- Takes a string `text`
- Returns a **dict** mapping each lowercased word to its count
- Ignores punctuation (`.`, `,`, `!`, `?`)

**Example:** `word_freq('Hello hello world!')` → `{'hello': 2, 'world': 1}`

```python
# ─── YOUR CODE HERE ───────────────────────────────────────────
from collections import Counter
import string
def word_freq(text: str) -> dict:
    translator = str.maketrans('', '', string.punctuation)
    textC = text.translate(translator)
    textC = textC.lower().split()
    counts = Counter(textC)
    return counts
# ──────────────────────────────────────────────────────────────
```

```python
# --- CHECK ---
assert word_freq('Hello hello world!') == {'hello': 2, 'world': 1}, 'Test 1 failed'
assert word_freq('a, a, b.') == {'a': 2, 'b': 1}, 'Test 2 failed'
assert word_freq('One') == {'one': 1}, 'Test 3 failed'
print('✅ All word_freq tests passed — 5/5 marks')
```

## Challenge 3 — Truthy Gate (5 marks)

Write a function `safe_divide(a, b)` that:
- Returns `a / b` if `b` is truthy (non-zero)
- Returns `None` if `b` is falsy, **without raising an exception**
- Uses short-circuit logic (no if-else allowed — use `and`/`or`)

```python
# ─── YOUR CODE HERE ───────────────────────────────────────────
def safe_divide(a, b):
    return (b and [a / b] or [None])[0]
# ──────────────────────────────────────────────────────────────
```

```python
# --- CHECK ---
assert safe_divide(10, 2) == 5.0
assert safe_divide(10, 0) is None
assert safe_divide(0, 5) == 0.0
assert safe_divide(7, 0.0) is None
print('✅ All safe_divide tests passed — 5/5 marks')
```

## Challenge 4 — Module Use: `random` & `string` (5 marks)

Write a function `random_password(length=12)` that generates a random password containing:
- Uppercase letters
- Lowercase letters  
- Digits
- At least one of each category

Use only `random` and `string` from the standard library.

```python
import random
import string

# ─── YOUR CODE HERE ───────────────────────────────────────────
def random_password(length: int = 12) -> str:
    chars = [
        random.choice(string.ascii_uppercase),
        random.choice(string.ascii_lowercase),
        random.choice(string.digits),
    ]
    all_chars = string.ascii_letters + string.digits
    chars += random.choices(all_chars, k=length - 3)
    random.shuffle(chars)
    return "".join(chars)
# ──────────────────────────────────────────────────────────────
```

```python
# --- CHECK ---
for _ in range(5):
    p = random_password()
    assert len(p) == 12, f'Expected length 12, got {len(p)}'
    assert any(c.isupper() for c in p), 'No uppercase'
    assert any(c.islower() for c in p), 'No lowercase'
    assert any(c.isdigit() for c in p), 'No digit'
print('✅ random_password tests passed — 5/5 marks')
print(p)
```

## Challenge 5 — CSV Row Parser (5 marks)

Write a function `parse_csv_row(line)` that:
- Takes a raw CSV line like `'  Alice , 30 , Engineer '`
- Returns a list of **stripped** values: `['Alice', '30', 'Engineer']`
- Handles any number of fields

```python
# ─── YOUR CODE HERE ───────────────────────────────────────────
def parse_csv_row(line: str) -> list:
    return [part.strip() for part in line.split(",")]
    
#parse_csv_row('  Alice , 30 , Engineer ')
# ──────────────────────────────────────────────────────────────
```

```python
# --- CHECK ---
assert parse_csv_row('  Alice , 30 , Engineer ') == ['Alice', '30', 'Engineer']
assert parse_csv_row('a,b,c') == ['a', 'b', 'c']
assert parse_csv_row(' x ') == ['x']
print('✅ All parse_csv_row tests passed — 5/5 marks')
```
