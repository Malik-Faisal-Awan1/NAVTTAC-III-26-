## Challenge 3 — Inheritance (5 marks)

Create:
- `Shape` base class with `area()` and `perimeter()` (raise `NotImplementedError`)
- `Circle(Shape)` with `radius`
- `Rectangle(Shape)` with `width` and `height`

Both subclasses must implement both methods.

```python
import math

# ─── YOUR CODE HERE ───────────────────────────────────────────
class Shape:
    pass

class Circle(Shape):
    pass

class Rectangle(Shape):
    pass
# ──────────────────────────────────────────────────────────────
```

```python
# --- CHECK ---
c = Circle(5)
assert abs(c.area() - math.pi * 25) < 0.001
assert abs(c.perimeter() - 2 * math.pi * 5) < 0.001
r = Rectangle(4, 6)
assert r.area() == 24
assert r.perimeter() == 20
print('✅ Shape hierarchy passed — 5/5 marks')
```
