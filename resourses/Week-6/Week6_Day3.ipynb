## Part A — Why This Topic Matters

In real data work, values often come in groups.

For example:

- marks of many students in many subjects
- daily temperature readings for many cities
- sales of many products over many months
- sensor values collected over time

We often need to do things like:

- add a bonus to all values
- subtract one row or one column from another
- find average values
- compare rows and columns
- scale data before machine learning

NumPy helps us do this quickly.

The biggest new idea today is **broadcasting**.
Broadcasting lets NumPy perform operations between arrays of different shapes when the shapes are still compatible.

```python
import numpy as np

np.set_printoptions(suppress=True)
print("NumPy imported successfully")
```

## Part B — Start with a Simple 1D Array

Let us start from something very easy.

We have marks of 5 students in one quiz.

```python
quiz_marks = np.array([60, 72, 81, 55, 90])
print("Quiz marks:", quiz_marks)
print("Shape:", quiz_marks.shape)
```

## Part C — Broadcasting with a Scalar

A **scalar** means one single value.

If we add a scalar to an array, NumPy adds that value to every element.
This is broadcasting in its simplest form.

```python
bonus_marks = quiz_marks + 5
print("Original marks:", quiz_marks)
print("After adding bonus of 5:", bonus_marks)
```

### What happened?

NumPy did not ask us to write a loop.
It automatically applied `+ 5` to all values.

This is faster and cleaner than manual looping.

```python
print("Subtract 10 from all values:", quiz_marks - 10)
print("Multiply all values by 2:", quiz_marks * 2)
print("Divide all values by 5:", quiz_marks / 5)
```

## Part D — Compare with Normal Python List

This is a useful comparison for students.

Python lists and NumPy arrays do not behave the same way.

```python
marks_list = [60, 72, 81, 55, 90]

print("Python list:", marks_list)
print("NumPy array:", quiz_marks)

try:
    print(marks_list + 5)
except Exception as e:
    print("List + 5 gives error:", e)

print("Array + 5 works:", quiz_marks + 5)
```

### Important idea

- Python list is a general container
- NumPy array is made for numerical computing
- broadcasting is one reason why NumPy is powerful

## Part E — Element-wise Operations Between Same-Shape Arrays

If two arrays have the same shape, NumPy applies operations element by element.

```python
class_test = np.array([10, 8, 9, 7, 10])
assignment = np.array([15, 14, 13, 12, 15])

print("Class test:", class_test)
print("Assignment:", assignment)
print("Total marks:", class_test + assignment)
print("Difference:", assignment - class_test)
print("Element-wise product:", class_test * assignment)
```

## Part F — A 2D Array Example

Now let us move to a more realistic case.

We will store marks of 4 students in 3 subjects.

Each row = one student  
Each column = one subject

```python
marks = np.array([
    [78, 85, 90],
    [67, 74, 80],
    [88, 92, 95],
    [59, 65, 70]
])

print("Marks table:")
print(marks)
print("Shape:", marks.shape)
```

## Part G — Broadcasting a Scalar on a 2D Array

Again, if we add one scalar value, NumPy adds it to every item in the full table.

```python
print("Original marks table:")
print(marks)

print("
Add 3 grace marks to every value:")
print(marks + 3)
```

## Part H — Broadcasting a 1D Array Across Rows

Suppose each subject has a different bonus:

- Subject 1 gets `+2`
- Subject 2 gets `+0`
- Subject 3 gets `+5`

Now we use a 1D array with 3 values.

```python
subject_bonus = np.array([2, 0, 5])
print("Subject bonus:", subject_bonus)
print("Shape of subject_bonus:", subject_bonus.shape)

updated_marks = marks + subject_bonus
print("
Updated marks after subject-wise bonus:")
print(updated_marks)
```

### Why did this work?

`marks` has shape `(4, 3)`  
`subject_bonus` has shape `(3,)`

NumPy matches the 3 bonus values with the 3 columns.
So the same bonus row is used for every student.

## Part I — Broadcasting a Column Vector Across Columns

Now suppose each student gets a different attendance bonus.

That means:

- student 1 gets `+1`
- student 2 gets `+2`
- student 3 gets `+0`
- student 4 gets `+3`

This is a row-wise idea, so we need a **column vector**.

```python
attendance_bonus = np.array([1, 2, 0, 3])
print("Attendance bonus:", attendance_bonus)
print("Shape before reshape:", attendance_bonus.shape)
```

### Why reshape is needed

If we keep shape `(4,)`, NumPy may not know that this should be treated as one value for each row.

So we convert it into shape `(4, 1)`.
This means 4 rows and 1 column.

```python
attendance_bonus_col = attendance_bonus.reshape(4, 1)
print("Reshaped attendance bonus:")
print(attendance_bonus_col)
print("Shape after reshape:", attendance_bonus_col.shape)
```

```python
new_marks = marks + attendance_bonus_col
print("Original marks:")
print(marks)

print("
Marks after student-wise attendance bonus:")
print(new_marks)
```

## Part J — Broadcasting Rule in Easy Words

NumPy compares shapes from the **right side**.

Two dimensions are compatible if:

- they are equal, or
- one of them is `1`

If this condition fails, broadcasting will not work.

Examples:

- `(4, 3)` and `(3,)` → works
- `(4, 3)` and `(4, 1)` → works
- `(4, 3)` and `(4,)` → does **not** work for column-wise matching
- `(4, 3)` and `(2,)` → does **not** work

## Part K — A Broadcasting Example That Fails

It is useful to show students one failing case too.

```python
wrong_bonus = np.array([1, 2])
print("Wrong bonus:", wrong_bonus)
print("Shape:", wrong_bonus.shape)

try:
    print(marks + wrong_bonus)
except Exception as e:
    print("Broadcasting error:", e)
```

### Why did it fail?

Our marks table has 3 columns.
But `wrong_bonus` has only 2 values.

NumPy cannot match 2 values with 3 columns.

## Part L — Another Easy Broadcasting Example

Let us make a small table and perform subtraction.

```python
temperatures = np.array([
    [30, 32, 31],
    [28, 29, 30],
    [35, 36, 34]
])

morning_base = np.array([1, 0, -1])

print("Temperature table:")
print(temperatures)
print("
Adjustment row:", morning_base)
print("
After broadcasting subtraction:")
print(temperatures - morning_base)
```

## Part M — The Axis Concept

Now we move to statistics.

This is where many students get confused.

### Easy meaning of axis in a 2D array

For a table like this:

- `axis=0` means go **down the rows** and operate column by column
- `axis=1` means go **across the columns** and operate row by row

A simple way to remember:

- `axis=0` gives result for each **column**
- `axis=1` gives result for each **row**

```python
print("Marks table:")
print(marks)
```

## Part N — Sum of All Values

If we do not give any axis, NumPy calculates over the full array.

```python
print("Total of all marks:", np.sum(marks))
```

## Part O — Column-wise Sum with `axis=0`

This tells us the total marks in each subject.

```python
print("Column-wise sum (subject totals):")
print(np.sum(marks, axis=0))
```

### Interpretation

The output has 3 values because we have 3 columns.
Each value is the total of one subject.

## Part P — Row-wise Sum with `axis=1`

This tells us total marks of each student.

```python
print("Row-wise sum (student totals):")
print(np.sum(marks, axis=1))
```

### Interpretation

The output has 4 values because we have 4 rows.
Each value is the total marks of one student.

## Part Q — Mean (Average)

We use `np.mean()` for average.

Again, we can do:

- overall mean
- column-wise mean
- row-wise mean

```python
print("Overall average:", np.mean(marks))
print("Subject-wise average:", np.mean(marks, axis=0))
print("Student-wise average:", np.mean(marks, axis=1))
```

## Part R — Minimum and Maximum

These help us see the smallest and largest values.

```python
print("Overall minimum:", np.min(marks))
print("Overall maximum:", np.max(marks))
print("Subject-wise minimum:", np.min(marks, axis=0))
print("Subject-wise maximum:", np.max(marks, axis=0))
```

## Part S — Standard Deviation

Standard deviation tells us how spread out the values are.

Easy idea:

- low standard deviation → values are closer together
- high standard deviation → values are more spread out

In machine learning and data analysis, this is very important.

```python
print("Overall standard deviation:", np.std(marks))
print("Subject-wise standard deviation:", np.std(marks, axis=0))
print("Student-wise standard deviation:", np.std(marks, axis=1))
```

## Part T — A Full Student Analysis Example

Now let us do a proper mini analysis.

We will calculate:

- total marks of each student
- average marks of each student
- average of each subject
- highest mark in each subject

```python
student_totals = np.sum(marks, axis=1)
student_averages = np.mean(marks, axis=1)
subject_averages = np.mean(marks, axis=0)
subject_highest = np.max(marks, axis=0)

print("Student totals:", student_totals)
print("Student averages:", student_averages)
print("Subject averages:", subject_averages)
print("Highest mark in each subject:", subject_highest)
```

## Part U — Keep Dimensions with `keepdims=True`

Sometimes we want the result to remain 2D.
This helps in later broadcasting.

```python
row_means = np.mean(marks, axis=1)
row_means_keep = np.mean(marks, axis=1, keepdims=True)

print("Row means without keepdims:")
print(row_means)
print("Shape:", row_means.shape)

print("
Row means with keepdims=True:")
print(row_means_keep)
print("Shape:", row_means_keep.shape)
```

### Why is this useful?

If we later want to subtract each student's own average from that student's row, keeping dimensions makes broadcasting easier.

```python
centered_marks = marks - np.mean(marks, axis=1, keepdims=True)
print("Marks after subtracting each student's own average:")
print(centered_marks)
```

## Part V — What is Normalization?

In simple words, normalization means changing data to a common scale.

Why do we do this?

Because some features may have very small values and others may have very large values.
This can affect machine learning models.

A common normalization method is **min-max scaling**.

## Part W — Min-Max Normalization Formula

The formula is:

\[
	ext{normalized} = rac{x - x_{min}}{x_{max} - x_{min}}
\]

This usually changes values into the range **0 to 1**.

```python
heights = np.array([150, 160, 170, 180, 190])

min_val = np.min(heights)
max_val = np.max(heights)
normalized_heights = (heights - min_val) / (max_val - min_val)

print("Original heights:", heights)
print("Normalized heights:", normalized_heights)
```

### Interpretation

- smallest value becomes `0`
- largest value becomes `1`
- values in between move proportionally between `0` and `1`

## Part X — Min-Max Normalization on a 2D Array Column-wise

In datasets, each column usually represents one feature.
So scaling is often done **column by column**.

```python
features = np.array([
    [150, 55, 30000],
    [165, 68, 45000],
    [172, 72, 52000],
    [180, 80, 61000]
], dtype=float)

print("Original feature table:")
print(features)
```

```python
col_min = np.min(features, axis=0)
col_max = np.max(features, axis=0)
scaled_features = (features - col_min) / (col_max - col_min)

print("Column minimums:", col_min)
print("Column maximums:", col_max)
print("
Scaled features (0 to 1):")
print(scaled_features)
```

### Why did this work?

- `features` has shape `(4, 3)`
- `col_min` has shape `(3,)`
- `col_max` has shape `(3,)`

NumPy uses broadcasting and applies each column's min and max to that column.

## Part Y — Standardization (Z-Score Style Scaling)

Another common scaling method is standardization.

Formula:

\[
	ext{standardized} = rac{x - \mu}{\sigma}
\]

Where:

- `\mu` is the mean
- `\sigma` is the standard deviation

This usually makes data centered around 0.

```python
col_mean = np.mean(features, axis=0)
col_std = np.std(features, axis=0)
standardized_features = (features - col_mean) / col_std

print("Column means:", col_mean)
print("Column std:", col_std)
print("
Standardized features:")
print(standardized_features)
```

## Part Z — A Practical Example: Temperature Analysis

Let us use a real-life style example.

Each row is one city.  
Each column is one day.

```python
city_temp = np.array([
    [30, 32, 31, 29],
    [25, 27, 26, 28],
    [35, 36, 34, 33]
])

print("Temperature table:")
print(city_temp)
```

```python
print("Average temperature of each city:", np.mean(city_temp, axis=1))
print("Average temperature of each day:", np.mean(city_temp, axis=0))
print("Maximum temperature in each city:", np.max(city_temp, axis=1))
print("Minimum temperature in each day:", np.min(city_temp, axis=0))
```

## Part AA — Broadcasting for Temperature Adjustment

Suppose a device error caused all day readings to need these corrections:

- Day 1: `+1`
- Day 2: `0`
- Day 3: `-1`
- Day 4: `+2`

```python
day_adjustment = np.array([1, 0, -1, 2])
corrected_temp = city_temp + day_adjustment

print("Day adjustment:", day_adjustment)
print("
Corrected temperature table:")
print(corrected_temp)
```

## Part AB — Centering Data by Column Mean

A useful analysis step is to subtract the mean of each column.
This tells us how far each value is from that column average.

```python
column_means = np.mean(city_temp, axis=0)
centered_by_day = city_temp - column_means

print("Column means:", column_means)
print("
Centered by day:")
print(centered_by_day)
```

## Part AC — Common Beginner Mistakes

### Mistake 1: Thinking broadcasting is random
It is not random. Shapes must be compatible.

### Mistake 2: Forgetting to compare shapes from the right side
Always check the last dimensions first.

### Mistake 3: Mixing `axis=0` and `axis=1`
- `axis=0` → column-wise result
- `axis=1` → row-wise result

### Mistake 4: Forgetting `reshape()` for column broadcasting
A row-like array `(4,)` is not the same as a column array `(4, 1)`.

### Mistake 5: Dividing integers and expecting special rounding
Scaling usually gives decimal values.

### Mistake 6: Applying normalization to the full table when you really need per-column scaling
In datasets, scaling is usually done feature by feature.
