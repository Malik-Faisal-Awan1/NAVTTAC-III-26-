## Part A — What is NumPy?

NumPy stands for **Numerical Python**.

It is a library used for working with numbers in a fast and organized way.

In normal Python, we often use lists.  
Lists are good, but when we want to do data science, machine learning, statistics, or scientific computing, lists become slow and limited.

NumPy gives us **arrays**.

An array is similar to a list, but it is more powerful for numerical work.

### Why NumPy is useful

- faster than normal Python lists for numeric operations
- easier to do calculations on many values at once
- important for Pandas, machine learning, and deep learning
- helps us think in a data-computing way

### Real-life example

Suppose you have marks of 500 students.

If you want to:
- add 5 grace marks to all students
- find average marks
- reshape data into rows and columns
- normalize the values

NumPy makes this much easier.

````python

import numpy as np

print("NumPy imported successfully")
````

## Part B — Python List vs NumPy Array

Let us first compare a simple list and a NumPy array.

A Python list can store mixed data types.  
A NumPy array usually stores values of one main type, which makes computation faster and cleaner.

````python

# Python list
marks_list = [60, 70, 80, 90]
print("Python list:", marks_list)
print("Type:", type(marks_list))

# NumPy array
marks_array = np.array([60, 70, 80, 90])
print("NumPy array:", marks_array)
print("Type:", type(marks_array))
````

### Explanation

In the code above:

- `np.array([...])` creates a NumPy array
- both look similar when printed
- but internally, NumPy arrays are designed for fast numerical work

Now let us see one very important difference.

````python

# Adding 5 to a list does not work directly
marks_list = [60, 70, 80, 90]

try:
    print(marks_list + 5)
except Exception as e:
    print("List error:", e)

# Adding 5 to a NumPy array works element by element
marks_array = np.array([60, 70, 80, 90])
print("Array + 5:", marks_array + 5)
````

### Important idea

For a Python list, `+` is used for joining lists, not adding a number to each item.

For a NumPy array, `+ 5` means:

- add 5 to first value
- add 5 to second value
- add 5 to third value
- and so on

This is one of the biggest reasons NumPy is powerful.

## Part C — Creating Arrays

We can create arrays in different ways.

````python

# 1D array
a = np.array([10, 20, 30, 40])
print("1D array:", a)

# 2D array
b = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print("2D array:")
print(b)

# Array of zeros
zeros_array = np.zeros((2, 3))
print("Zeros array:")
print(zeros_array)

# Array of ones
ones_array = np.ones((3, 2))
print("Ones array:")
print(ones_array)

# Identity matrix
identity_array = np.eye(3)
print("Identity matrix:")
print(identity_array)
````

### Explanation

- `np.array()` creates an array from given values
- `np.zeros((rows, cols))` creates an array filled with 0
- `np.ones((rows, cols))` creates an array filled with 1
- `np.eye(n)` creates an identity matrix

### What is an identity matrix?

It is a square matrix where:

- diagonal values are 1
- all other values are 0

Example:

````python

example_identity = np.eye(4)
print(example_identity)
````

## Part D — dtype, shape, size, ndim

These are very important properties of an array.

- **dtype** = data type of values
- **shape** = structure of the array
- **size** = total number of values
- **ndim** = number of dimensions

````python

arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print("Array:")
print(arr)

print("dtype:", arr.dtype)
print("shape:", arr.shape)
print("size:", arr.size)
print("ndim:", arr.ndim)
````

### Explanation

For this array:

```python
[
 [10, 20, 30],
 [40, 50, 60]
]
```

- `shape` is `(2, 3)` because there are 2 rows and 3 columns
- `size` is `6` because total values are 6
- `ndim` is `2` because it is a 2D array
- `dtype` tells the kind of values stored in the array

Let us also see how NumPy handles mixed values.

````python

mixed = np.array([1, 2, 3.5])
print("Mixed array:", mixed)
print("dtype:", mixed.dtype)
````

### Important note

NumPy tries to keep one common data type in the whole array.

Here we used integers and a float together, so NumPy converted all values to float.

That is why output dtype becomes something like `float64`.

## Part E — Reshape

Sometimes data comes in one long line, but we want it in rows and columns.

For this, we use `reshape()`.

````python

numbers = np.array([1, 2, 3, 4, 5, 6])
print("Original array:", numbers)
print("Original shape:", numbers.shape)

reshaped = numbers.reshape(2, 3)
print("Reshaped array:")
print(reshaped)
print("New shape:", reshaped.shape)
````

### Explanation

The original array has 6 values.

When we write:

```python
numbers.reshape(2, 3)
```

it means:

- make 2 rows
- make 3 columns

Since `2 x 3 = 6`, reshape works.

### Very important rule

The total number of values must stay the same.

So:

- 6 values can become `(2, 3)`
- 6 values can become `(3, 2)`
- 6 values cannot become `(4, 2)`

because `4 x 2 = 8`, and we do not have 8 values.

````python

numbers = np.array([1, 2, 3, 4, 5, 6])

print("Reshape to 3 x 2:")
print(numbers.reshape(3, 2))

# Automatic dimension using -1
print("Reshape with -1:")
print(numbers.reshape(2, -1))
````

### About `-1`

When we use `-1`, NumPy automatically calculates that dimension.

Example:

```python
numbers.reshape(2, -1)
```

We already fixed rows as 2.  
There are 6 values total.  
So NumPy understands that columns must be 3.

## Part F — arange() and linspace()

These two functions are very common.

### `arange()`

It works like Python `range()`, but it creates a NumPy array.

### `linspace()`

It creates evenly spaced values between a start and end point.

````python

# arange examples
print("arange(5):", np.arange(5))
print("arange(2, 10):", np.arange(2, 10))
print("arange(2, 11, 2):", np.arange(2, 11, 2))
````

### Explanation of `arange()`

- `np.arange(5)` gives values from 0 to 4
- `np.arange(2, 10)` gives values from 2 to 9
- `np.arange(2, 11, 2)` gives values with step 2

So it follows this pattern:

```python
np.arange(start, stop, step)
```

The `stop` value is not included.

````python

# linspace examples
print("linspace(0, 1, 5):", np.linspace(0, 1, 5))
print("linspace(10, 20, 6):", np.linspace(10, 20, 6))
````

### Explanation of `linspace()`

`np.linspace(start, end, number_of_values)`

It includes both start and end values.

Example:

```python
np.linspace(0, 1, 5)
```

This creates 5 equally spaced values from 0 to 1.

This is useful in graphs, mathematics, simulations, and sampling values.

## Part G — Basic Operations on Arrays

One big strength of NumPy is that operations happen element by element.

````python

x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 30, 40])

print("x:", x)
print("y:", y)

print("x + y:", x + y)
print("x - y:", x - y)
print("x * y:", x * y)
print("x / y:", x / y)
````

### Explanation

When we do:

- `x + y` → first values add, second values add, and so on
- `x - y` → subtraction happens element by element
- `x * y` → multiplication happens element by element
- `x / y` → division happens element by element

This is much easier than using loops for every small job.

````python

arr = np.array([5, 10, 15, 20])

print("Original:", arr)
print("arr + 2:", arr + 2)
print("arr * 3:", arr * 3)
print("arr ** 2:", arr ** 2)
````

### Explanation

Here we used one array with one number.

NumPy automatically applies that operation to every element.

So:

- `arr + 2` means add 2 to every value
- `arr * 3` means multiply every value by 3
- `arr ** 2` means square every value

## Part H — Vectorization Mindset

This is one of the most important concepts of today.

### What is vectorization?

Vectorization means:

**doing operations on the whole array at once, instead of using manual loops**

This makes code:

- shorter
- cleaner
- faster
- easier to read

Let us compare loop-based thinking and NumPy thinking.

````python

# Normal Python loop style
values = [1, 2, 3, 4, 5]
result_loop = []

for item in values:
    result_loop.append(item * 2)

print("Using loop:", result_loop)

# NumPy vectorized style
values_np = np.array([1, 2, 3, 4, 5])
result_vector = values_np * 2

print("Using NumPy vectorization:", result_vector)
````

### Why vectorization is better

In the loop:

- we create an empty list
- go through values one by one
- manually multiply
- manually append

In the NumPy version:

- one line does the work on all values

This is the NumPy mindset.

### Real-life thinking

Suppose you have salaries of 10,000 employees and want to increase all by 10%.

With normal loop thinking, you update one by one.  
With NumPy, you can do the whole array in one step.

````python

salaries = np.array([30000, 45000, 50000, 60000])
new_salaries = salaries * 1.10

print("Old salaries:", salaries)
print("New salaries after 10% increase:", new_salaries)
````
