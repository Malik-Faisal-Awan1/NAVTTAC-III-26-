## Part B — First Understand the Words

### Vector
A **vector** is a 1D array.

Example:
- `[2, 4, 6]`

### Matrix
A **matrix** is a 2D array.

Example:
- `[[1, 2], [3, 4]]`

In NumPy:
- 1D array can act like a vector
- 2D array can act like a matrix

```python
vector = np.array([2, 4, 6])
matrix = np.array([
    [1, 2],
    [3, 4]
])

print("Vector:")
print(vector)
print("Shape:", vector.shape)

print("\nMatrix:")
print(matrix)
print("Shape:", matrix.shape)
```

## Part C — Basic Operations on Vectors

NumPy allows element-wise operations on vectors of the same shape.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("a:", a)
print("b:", b)
print("\na + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
```

### Important note

The operation `a * b` above is **element-wise multiplication**.

It is **not** the dot product.

This is a very common beginner confusion.

## Part D — Dot Product of Two Vectors

The **dot product** multiplies matching values and then adds the results.

For example:

If  
`a = [1, 2, 3]`  
`b = [4, 5, 6]`

Then:

\[
(1 \times 4) + (2 \times 5) + (3 \times 6) = 4 + 10 + 18 = 32
\]

This gives one final number.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

dot_result = np.dot(a, b)

print("a:", a)
print("b:", b)
print("Dot product using np.dot(a, b):", dot_result)
```

## Part E — Manual Check of Dot Product

Let us verify the answer manually so students understand the logic clearly.

```python
manual_dot = (1 * 4) + (2 * 5) + (3 * 6)
print("Manual dot product:", manual_dot)
```

## Part F — A Real-Life Style Dot Product Example

Suppose a student gets these marks in 3 quizzes:

- Quiz 1 = 80
- Quiz 2 = 70
- Quiz 3 = 90

And the quiz weights are:

- 20%
- 30%
- 50%

Then final weighted score can be found using dot product.

```python
marks = np.array([80, 70, 90])
weights = np.array([0.2, 0.3, 0.5])

weighted_score = np.dot(marks, weights)

print("Marks:", marks)
print("Weights:", weights)
print("Weighted score:", weighted_score)
```

## Part G — 2D Arrays as Tables

Now let us move to matrices.

Suppose 3 students have marks in 3 subjects.

- each row = one student
- each column = one subject

```python
student_marks = np.array([
    [78, 85, 90],
    [67, 74, 80],
    [88, 92, 95]
])

print("Student marks matrix:")
print(student_marks)
print("Shape:", student_marks.shape)
```

## Part H — Element-Wise Matrix Operations

When two matrices have the same shape, operations happen position by position.

```python
bonus = np.array([
    [2, 2, 2],
    [3, 3, 3],
    [1, 1, 1]
])

print("Original marks:")
print(student_marks)

print("\nBonus matrix:")
print(bonus)

print("\nAfter element-wise addition:")
print(student_marks + bonus)
```

## Part I — Scalar Operations on a Matrix

A scalar operation affects every value in the matrix.

```python
print("Add 5 to every value:")
print(student_marks + 5)

print("\nMultiply every value by 2:")
print(student_marks * 2)
```

## Part J — Transpose of a Matrix

The **transpose** changes rows into columns.

If original shape is `(rows, columns)`, transpose becomes `(columns, rows)`.

In NumPy, transpose can be written as:

- `matrix.T`
- `np.transpose(matrix)`

```python
print("Original matrix:")
print(student_marks)
print("Shape:", student_marks.shape)

transposed = student_marks.T
print("\nTransposed matrix:")
print(transposed)
print("Shape:", transposed.shape)
```

### Why transpose matters

Transpose is useful when:

- changing data layout
- preparing data for matrix multiplication
- working with machine learning formulas
- comparing row-based and column-based information

## Part K — Matrix Multiplication

This is different from element-wise multiplication.

In NumPy, matrix multiplication can be done using:

- `A @ B`
- `np.dot(A, B)` for suitable shapes

For matrix multiplication:

- the number of columns in the first matrix
- must match the number of rows in the second matrix

```python
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

print("A:")
print(A)
print("\nB:")
print(B)

print("\nElement-wise multiplication A * B:")
print(A * B)

print("\nMatrix multiplication A @ B:")
print(A @ B)
```

## Part L — Understand One Entry of Matrix Multiplication

Let us understand the top-left value of `A @ B`.

Using:

\[
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
\quad
B =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
\]

Top-left value:

\[
(1 \times 5) + (2 \times 7) = 5 + 14 = 19
\]

That is why matrix multiplication is based on row-by-column work.

```python
result = A @ B
print("A @ B =")
print(result)

print("\nTop-left value:", result[0, 0])
```

## Part M — Matrix and Vector Multiplication

A matrix can also multiply a vector if shapes match.

```python
M = np.array([
    [2, 1, 0],
    [1, 3, 2]
])

v = np.array([4, 5, 6])

print("Matrix M:")
print(M)
print("Shape:", M.shape)

print("\nVector v:")
print(v)
print("Shape:", v.shape)

print("\nM @ v =")
print(M @ v)
```

## Part N — Identity Matrix

An **identity matrix** is like the number 1 for matrix multiplication.

For a 3×3 identity matrix:

\[
I =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\]

Multiplying a matrix by the identity matrix keeps it unchanged.

```python
I = np.eye(3)

print("Identity matrix I:")
print(I)

print("\nstudent_marks @ I =")
print(student_marks @ I)
```

## Part O — Determinant (Basic Idea)

The **determinant** is a special value for square matrices.

For beginners, the important idea is:

- determinant is defined for square matrices
- if determinant is `0`, the matrix is singular
- singular matrix does not have an inverse

```python
C = np.array([
    [2, 1],
    [5, 3]
])

det_C = np.linalg.det(C)

print("Matrix C:")
print(C)
print("\nDeterminant of C:", det_C)
```

## Part P — Inverse of a Matrix

The inverse of a matrix is a special matrix that can undo the original one.

If `C_inv` is the inverse of `C`, then:

\[
C \cdot C^{-1} = I
\]

But not every matrix has an inverse.

```python
C_inv = np.linalg.inv(C)

print("Matrix C:")
print(C)

print("\nInverse of C:")
print(C_inv)

print("\nC @ C_inv =")
print(C @ C_inv)
```

### Important note

Because of decimal rounding, the result may not look perfectly exact.

You may see values like `1.0000000` or very tiny values close to `0`.

That is normal in numerical computing.

## Part Q — Solve a Simple Linear System

Suppose we want to solve:

\[
2x + y = 11
\]
\[
5x + 3y = 29
\]

We can write it in matrix form and solve it using `np.linalg.solve()`.

```python
A_eq = np.array([
    [2, 1],
    [5, 3]
])

b_eq = np.array([11, 29])

solution = np.linalg.solve(A_eq, b_eq)

print("Coefficient matrix A_eq:")
print(A_eq)

print("\nAnswer vector b_eq:")
print(b_eq)

print("\nSolution [x, y]:")
print(solution)
```

## Part R — Check the Solution

Let us verify the values by putting them back into the equations.

```python
x, y = solution

print("x =", x)
print("y =", y)

print("\nCheck equation 1: 2x + y =", 2*x + y)
print("Check equation 2: 5x + 3y =", 5*x + 3*y)
```

## Part S — Another Useful Operation: Sum by Rows and Columns

Even when working with matrices in linear algebra, basic row and column summaries are still important.

```python
sales = np.array([
    [100, 120, 130],
    [90, 110, 140],
    [150, 160, 170]
])

print("Sales matrix:")
print(sales)

print("\nTotal of each row:", np.sum(sales, axis=1))
print("Total of each column:", np.sum(sales, axis=0))
```

## Part X — Random Choice from Given Values

Use `np.random.choice()` when you want random selection from specific options.

```python
colors = np.array(["red", "blue", "green", "yellow"])

chosen_colors = np.random.choice(colors, size=6)

print("Available colors:", colors)
print("Random choices:", chosen_colors)
```

## Part AA — Why Results Change Every Time

If you run random code again and again, you usually get different results.

That is normal.

But in teaching, testing, and machine learning experiments, sometimes we want the same random result again.

That is where a **seed** helps.

## Part AB — Reproducibility with `np.random.seed()`

A seed fixes the random starting point.

If we use the same seed and same code, we get the same result.

```python
np.random.seed(42)
sample1 = np.random.randint(1, 100, size=5)

np.random.seed(42)
sample2 = np.random.randint(1, 100, size=5)

print("Sample 1:", sample1)
print("Sample 2:", sample2)
print("Are both equal?", np.array_equal(sample1, sample2))
```

## Part AC — Change the Seed, Change the Result

Different seed usually gives a different output.

```python
np.random.seed(1)
a1 = np.random.randint(1, 20, size=5)

np.random.seed(7)
a2 = np.random.randint(1, 20, size=5)

print("Seed 1 result:", a1)
print("Seed 7 result:", a2)
```

## Part AG — Performance Intuition

Now let us understand one big reason why NumPy is popular.

NumPy is usually much faster than manual Python loops for numerical arrays.

Why?

Because NumPy operations are implemented in optimized low-level code.

We do not need deep theory right now. We just need the practical idea:

- loops are slower for large numeric work
- vectorized NumPy operations are usually faster and cleaner

## Part AH — Simple Timing Example: Python Loop

We will add `1` to a large list using normal Python loop style.

```python
numbers_list = list(range(200000))

start = time.time()
result_list = [x + 1 for x in numbers_list]
end = time.time()

print("Python list operation finished")
print("Time taken:", end - start, "seconds")
```

## Part AI — Simple Timing Example: NumPy Vectorized Operation

Now we do a similar operation with NumPy.

```python
numbers_array = np.arange(200000)

start = time.time()
result_array = numbers_array + 1
end = time.time()

print("NumPy array operation finished")
print("Time taken:", end - start, "seconds")
```

## Part AJ — What Should Students Notice?

Students should notice two things:

1. NumPy code is shorter and cleaner
2. NumPy is usually faster for large numeric work

This is called the **vectorization mindset**:

Instead of processing one element at a time with loops, let NumPy process the full array at once.

## Part AL — Common Beginner Mistakes

### Mistake 1: Mixing up `*` and `@`
- `*` means element-wise multiplication
- `@` means matrix multiplication

### Mistake 2: Forgetting shape rules
For multiplication, shapes must match properly.

### Mistake 3: Thinking random values should stay the same
They will change unless you use a seed.

### Mistake 4: Using inverse everywhere
For beginners, use inverse only when needed.  
For solving equations, `np.linalg.solve()` is often a better direct choice.

### Mistake 5: Ignoring decimal rounding
Numerical results may show tiny rounding differences. That is normal.
