## Part A — Dictionary Basics (Key → Value)

### Example A1: Create dictionaries

```python
student = {"name": "Ali", "marks": 92, "city": "Lahore"}
print(student)
print(type(student))
```

```python
d = {} # Empty dictionary
print(d)
```

### Example A2: Access values (and KeyError problem)les are immutable

```python
student = {"name": "Ali", "marks": 92}

print(student["name"])   # OK
# print(student["age"])  # KeyError (not present)
```

```python
print(student.get("age"))           # None
print(student.get("age", "N/A"))    # N/A
```

### Example A3: Update, add, delete

```python
student = {"name": "Maria", "marks": 92}

student["marks"] = 95      # update
student["grade"] = "A"     # add new key

print(student)

del student["grade"]       # delete key
print(student)
```

```python
# Delete safely
student.pop("grade", None)  # no error if missing
```

## Part B — Looping Through Dictionaries

### Example B1: Loop keys

```python
student = {"name": "Maria", "marks": 92, "city": "Lahore"}

for k in student:
    print(k)        # keys by default
```

### Example B2: keys(), values(), items()

```python
for k in student.keys():
    print("Key:", k)

for v in student.values():
    print("Value:", v)

for k, v in student.items():
    print(k, "=>", v)
```

## Part C — Build a Dictionary from User Input

```python
record = {}

record["name"] = input("Name: ").strip()
record["roll"] = input("Roll: ").strip()

marks = input("Marks (0-100): ").strip()
if marks.isdigit() and 0 <= int(marks) <= 100:
    record["marks"] = int(marks)
else:
    record["marks"] = None

print(record)
```

## Part D — Frequency Counting

### Example D1: Count word frequency (using get)

```python
text = "python is easy and python is powerful"
words = text.lower().split()

freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1

print(freq)
```

```python
for word, count in freq.items():
    print(f"{word:<10} : {count}")
```

### Example D2: Character frequency (ignore spaces)

```python
s = input("Enter a sentence: ")

count = {}
for ch in s.lower():
    if ch == " ":
        continue
    count[ch] = count.get(ch, 0) + 1

print(count)
```

## Part E — Grouping / Categorizing using Dict

```python
students = [
    ("Maria", 92),
    ("Ali", 78),
    ("Zain", 88),
    ("Haseeb", 61),
    ("Sara", 95)
]

groups = {"A": [], "B": [], "C": [], "D": []}

for name, marks in students:
    if marks >= 90:
        groups["A"].append(name)
    elif marks >= 80:
        groups["B"].append(name)
    elif marks >= 70:
        groups["C"].append(name)
    else:
        groups["D"].append(name)

print(groups)
```

```python
for grade, names in groups.items():
    print(grade, ":", names)
```

## Part F — Nested Dictionaries

```python
students = {
    101: {"name": "Maria", "marks": 92, "city": "Lahore"},
    102: {"name": "Ali", "marks": 78, "city": "Islamabad"},
    103: {"name": "Zain", "marks": 88, "city": "Peshawar"},
}

print(students[101]["name"])   # Maria
print(students[102]["marks"])  # 78
```

```python
for roll, info in students.items():
    print(roll, "=>", info["name"], info["marks"])
```
