# **Week 3 Day 2: Tuples and Sets**

## Part A — Tuples (Fixed / Read-only collection)

### Example A1: Creating tuples

```python
t1 = (10, 20, 30)
t2 = "Fraz", 22, "Lahore"   # parentheses optional
t3 = ()                      # empty tuple

print(t1)
print(t2)
print(type(t2))
```

```python
# Single item tuple (common mistake)
a = (5)     # not a tuple, it's int
b = (5,)    # tuple

print(type(a))
print(type(b))
```

### Example A2: Tuples are immutable

```python
data = (1, 2, 3)
# data[0] = 99  # Error (cannot change)
print(data[0])
```

```python
# But a tuple can contain a list (list is still mutable):
t = (1, [10, 20], 3)
t[1].append(99)
print(t)  # (1, [10, 20, 99], 3)
```

### Example A3: Tuple indexing and slicing

```python
t = ("a", "b", "c", "d", "e")
print(t[0])
print(t[-1])
print(t[1:4])
```

## Part B — Packing & Unpacking

### Example B1: Packing (automatic tuple)

```python
record = "Fraz", 92, "A"
print(record)
```

### Example B2: Unpacking

```python
name, marks, grade = record
print(name)
print(marks)
print(grade)
```

### Example B3: Swapping values

```python
a = 10
b = 20
a, b = b, a
print(a, b)
```

### Example B4: Unpacking with * (collect remaining)

```python
nums = (1, 2, 3, 4, 5)
first, *middle, last = nums
print(first)    # 1
print(middle)   # [2, 3, 4]
print(last)     # 5
```

## Part C — Tuples in Real Use

### Example C1: Returning multiple values from a function

```python
def calc(a, b):
    return a + b, a - b, a * b

add, sub, mul = calc(10, 5)
print(add, sub, mul)
```

### Example C2: Tuple as a “fixed record”

```python
# Each student is a tuple: (roll, name, marks)
students = [
    (101, "Maria", 92),
    (102, "Ali", 78),
    (103, "Zain", 88)
]

for roll, name, marks in students:
    print(roll, name, marks)
```

```python
# Find topper
top = students[0]
for s in students:
    if s[2] > top[2]:
        top = s

print("Topper:", top)
```

## Part D — Sets (Unique items, no duplicates)

### Example D1: Create sets

```python
a = {1, 2, 3, 3, 2}
print(a)   # duplicates removed

empty_set = set()  # not {}
print(type(empty_set))
```

### Example D2: Membership test (FAST)

```python
allowed = {"admin", "teacher", "student"}

role = input("Enter role: ").strip().lower()
if role in allowed:
    print("Allowed")
else:
    print("Not allowed")
```

### Example D3: Add / remove

```python
s = {1, 2, 3}
s.add(4)
print(s)

s.remove(2)   # error if not present
print(s)

s.discard(10) # no error if not present
print(s)
```

### Example D4: Remove duplicates from a list (two ways)

```python
# Way 1 (simple, order not guaranteed)
nums = [5, 2, 5, 1, 2, 9]
unique = list(set(nums))
print(unique)  # order may change
```

```python
# Way 2 (keep order)
nums = [5, 2, 5, 1, 2, 9]
seen = set()
unique = []

for x in nums:
    if x not in seen:
        seen.add(x)
        unique.append(x)

print(unique)  # [5,2,1,9]
```

## Part E — Set Operations

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("Union:", A | B)            # all
print("Intersection:", A & B)     # common
print("Difference:", A - B)       # in A not in B
print("Symmetric diff:", A ^ B)   # in one only
```

## Part F — Real Problems with Sets

### Example F1: Common subjects between two students

```python
maria = {"math", "english", "cs"}
ali = {"cs", "physics", "math"}

common = maria & ali
print(common)  # {'math','cs'}
```

### Example F2: Unique words in a sentence

```python
text = "Python is easy and Python is powerful"
words = text.lower().split()

unique_words = set(words)
print("Unique count:", len(unique_words))
print(unique_words)
```
