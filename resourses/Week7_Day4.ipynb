## Part A — Why Visualization Matters

When we work with data, raw rows are useful, but they are not always easy to understand quickly.

A graph helps us see:

- which category is bigger
- whether values are increasing or decreasing
- whether data is spread out or tightly grouped
- whether two variables move together
- whether there may be unusual values

A good graph helps us explain data to other people in a much easier way.

## Part E — Monthly Sales Summary for Plotting

Before making some plots, we first create grouped summaries.


```python
monthly_sales = sales.groupby("month", as_index=False)["sales"].sum()
monthly_sales
```

## Part F — Matplotlib Line Plot

A **line plot** is useful when:

- data is ordered
- we want to show change over time
- we want to see trend, rise, or fall

Here we will show total sales month by month.


```python
plt.figure(figsize=(8, 4))
plt.plot(monthly_sales["month"], monthly_sales["sales"], marker="o")
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

### What does this graph show?

This graph helps us answer:

- are sales increasing or decreasing?
- which month has the highest sales?
- which month has the lowest sales?

A line plot is best when the x-axis has order, such as days, months, or years.


## Part G — Improve Readability of a Line Plot

A graph should not only exist. It should also be easy to read.


```python
plt.figure(figsize=(9, 4))
plt.plot(monthly_sales["month"], monthly_sales["sales"], marker="o", linewidth=2)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Part H — City Sales Summary


```python
city_sales = sales.groupby("city", as_index=False)["sales"].sum()
city_sales
```

## Part I — Matplotlib Bar Plot

A **bar plot** is useful when:

- we compare categories
- we want to compare totals, averages, or counts
- the x-axis is not time-based order

Here we compare total sales across cities.


```python
plt.figure(figsize=(8, 4))
plt.bar(city_sales["city"], city_sales["sales"])
plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Sales")
plt.show()
```

### What can we interpret?

From a bar plot, we can compare heights.

Higher bar means larger value.

This helps us quickly answer:

- which city performed best?
- which city had lower sales?
- are the differences small or large?


## Part J — Bar Plot for Category Average Profit


```python
category_profit = sales.groupby("category", as_index=False)["profit"].mean()
category_profit
```

```python
plt.figure(figsize=(8, 4))
plt.bar(category_profit["category"], category_profit["profit"])
plt.title("Average Profit by Category")
plt.xlabel("Category")
plt.ylabel("Average Profit")
plt.xticks(rotation=15)
plt.show()
```

## Part K — Horizontal Bar Plot

Sometimes category names are long. In that case, a horizontal bar plot is easier to read.


```python
plt.figure(figsize=(8, 4))
plt.barh(category_profit["category"], category_profit["profit"])
plt.title("Average Profit by Category")
plt.xlabel("Average Profit")
plt.ylabel("Category")
plt.show()
```

## Part L — Matplotlib Histogram

A **histogram** is useful when:

- we want to see distribution of one numeric column
- we want to know whether values are spread out
- we want to see whether many values fall in one range

Let us see the distribution of `sales`.


```python
plt.figure(figsize=(8, 4))
plt.hist(sales["sales"], bins=6)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

### What does a histogram tell us?

It does **not** compare categories.

It groups numeric values into ranges called bins.

It helps us answer:

- are most sales values low, medium, or high?
- are values balanced or concentrated in one range?
- are there possible unusual values?


## Part M — Change Number of Bins

The number of bins changes the look of the histogram.


```python
plt.figure(figsize=(8, 4))
plt.hist(sales["sales"], bins=10)
plt.title("Distribution of Sales with More Bins")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

### Important idea

Too few bins can hide detail.  
Too many bins can make the graph noisy.

So the graph should be chosen carefully.


## Part N — Matplotlib Scatter Plot

A **scatter plot** is useful when:

- we compare two numeric variables
- we want to see relationship or trend
- we want to notice clusters or unusual points

Now we compare `sales` and `profit`.


```python
plt.figure(figsize=(8, 4))
plt.scatter(sales["sales"], sales["profit"])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

### What can we read from a scatter plot?

We try to answer:

- when sales increase, does profit also increase?
- do points move upward together?
- is there any point far away from the others?

If points show an upward pattern, there may be a positive relationship.


## Part O — Scatter Plot with Another Pair of Variables


```python
plt.figure(figsize=(8, 4))
plt.scatter(sales["discount"], sales["profit"])
plt.title("Discount vs Profit")
plt.xlabel("Discount (%)")
plt.ylabel("Profit")
plt.show()
```

### Think about this result

If discounts become larger, profit may decrease in many business cases.

A scatter plot can help us inspect that idea visually.

## Part Q — Seaborn Countplot

A **countplot** is useful when:

- we want to count how many records belong to each category
- we do not want sums or averages
- we only want frequency of labels

Let us count orders by city.


```python
plt.figure(figsize=(8, 4))
sns.countplot(data=sales, x="city")
plt.title("Number of Orders by City")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

### Countplot vs Bar Plot

This is a very important beginner point:

- **countplot** counts rows automatically
- **bar plot** usually shows a value that we already calculated, such as total sales or average profit

So they are not the same thing.


## Part R — Countplot by Category


```python
plt.figure(figsize=(8, 4))
sns.countplot(data=sales, x="category")
plt.title("Number of Records by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=15)
plt.show()
```

## Part S — Seaborn Boxplot

A **boxplot** is useful when:

- we want to compare the distribution of numeric data across categories
- we want to notice median, spread, and possible outliers
- we want more information than a bar plot

Let us compare `sales` across categories.


```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=sales, x="category", y="sales")
plt.title("Sales Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=15)
plt.show()
```

### Simple reading of a boxplot

A boxplot helps us notice:

- the center of the values
- how spread out the values are
- whether one category has more variation than another
- whether there are points far away from the usual range

This is useful when averages alone are not enough.


## Part T — Boxplot for Profit by City


```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=sales, x="city", y="profit")
plt.title("Profit Distribution by City")
plt.xlabel("City")
plt.ylabel("Profit")
plt.show()
```

## Part U — Seaborn Violin Plot

A **violin plot** is similar to a boxplot, but it also gives a better idea of shape and density.

It is useful when:

- we want to compare distributions
- we want to see where values are more concentrated


```python
plt.figure(figsize=(8, 4))
sns.violinplot(data=sales, x="category", y="rating")
plt.title("Rating Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Rating")
plt.xticks(rotation=15)
plt.show()
```

### Boxplot vs Violin Plot

- **boxplot** is simpler and easier for beginners
- **violin plot** gives a richer picture of data spread

If the class is new to plotting, start with boxplot first, then show violin plot.


## Part V — Heatmap Using Correlation

A **heatmap** is useful when:

- we want to compare many numeric relationships together
- we want to see which variables move together
- we want a quick summary matrix

First we compute correlation for numeric columns.


```python
numeric_cols = sales[["units", "sales", "profit", "discount", "rating"]]
corr_matrix = numeric_cols.corr()
corr_matrix
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### What does the heatmap show?

Each cell shows how strongly two numeric variables are related.

General idea:

- value close to **1** means strong positive relation
- value close to **-1** means strong negative relation
- value close to **0** means weak or no clear linear relation

This does not prove cause. It only shows association.


## Part W — Heatmap from a Pivot Table

Heatmaps are also useful on pivot tables.

Let us create city-category sales totals and view them as a heatmap.


```python
city_category_sales = sales.pivot_table(
    values="sales",
    index="city",
    columns="category",
    aggfunc="sum"
)

city_category_sales
```

```python
plt.figure(figsize=(8, 4))
sns.heatmap(city_category_sales, annot=True, cmap="YlGnBu")
plt.title("City and Category Sales Heatmap")
plt.xlabel("Category")
plt.ylabel("City")
plt.show()
```

## Part X — Pair of Good Questions Before Choosing a Plot

Before plotting, ask:

1. **What kind of columns do I have?**
   - categorical?
   - numeric?
   - time-based?

2. **What do I want to show?**
   - trend?
   - comparison?
   - distribution?
   - relationship?
   - frequency?

These two questions help you choose the correct graph.


## Part Y — Quick Plot Selection Guide

Use this simple rule:

- **Line plot** → trend over time  
- **Bar plot** → compare category totals or averages  
- **Histogram** → distribution of one numeric column  
- **Scatter plot** → relationship between two numeric columns  
- **Countplot** → count records in each category  
- **Boxplot** → compare spread of numeric data across categories  
- **Violin plot** → compare distribution shape across categories  
- **Heatmap** → compare many values in a matrix or correlation table

## Part AG — Common Beginner Mistakes

Some very common mistakes are:

- choosing a line plot for categories that have no real order
- using a bar plot when a countplot is actually needed
- forgetting axis labels
- creating a graph but not explaining the pattern
- using too many bins or too few bins in a histogram
- comparing raw values when grouped values were required
- reading correlation as proof of cause
