## Part A — Why Functions?

Functions help you:
- avoid repeating code
- make code readable
- test small parts easily
- reuse code in multiple programs

## Part B — Function Basics: Syntax + Parameters/Arguments

### Example B1: Small function (no input)

```python
def greet():
    print("Hello! Welcome to Python.")

greet()
greet()
```

### Example B2: Function with parameters (input to function)

```python
def greet_user(name):
    print(f"Hello, {name}!")

greet_user("Umer")
greet_user("Ali")
```

## Part C — Return Values

### Example C1: Return vs print (common confusion)

```python
def add_print(a, b):
    print(a + b)

def add_return(a, b):
    return a + b

x = add_print(2, 3)
print("x =", x)   # None, because function didn't return

y = add_return(2, 3)
print("y =", y)   # 5
```

### Example C2: Multiple returns (decision-based)

```python
def sign(n):
    if n > 0:
        return "positive"
    if n < 0:
        return "negative"
    return "zero"

print(sign(10))
print(sign(-5))
print(sign(0))
```

### Example C3: Return multiple values (tuple return)

```python
def calc(a, b):
    return a + b, a - b, a * b

add, sub, mul = calc(10, 5)
print(add, sub, mul)
```

## Part D — Default values

```python
def power(base, exp=2):
    return base ** exp

print(power(5))     # 25
print(power(5, 3))  # 125
```

## Part E — Decomposition (Break problem into functions)

### Example E1: Clean name function (reuse)

```python
def clean_name(name):
    name = name.strip()
    name = " ".join(name.split())
    return name.title()

print(clean_name("   fRaZ    asLam  "))
```

### Example E2: Validate marks function

```python
def is_valid_marks(s):
    return s.isdigit() and 0 <= int(s) <= 100

print(is_valid_marks("92"))
print(is_valid_marks("200"))
print(is_valid_marks("abc"))
```

### Example E3: Build a student record using functions

```python
def clean_name(name):
    name = name.strip()
    name = " ".join(name.split())
    return name.title()

def is_valid_marks(s):
    return s.isdigit() and 0 <= int(s) <= 100

def create_student():
    name = clean_name(input("Name: "))
    marks_s = input("Marks (0-100): ").strip()
    if not is_valid_marks(marks_s):
        return None  # invalid record
    return {"name": name, "marks": int(marks_s)}

student = create_student()
print(student)
```

## Part F — A Common Pattern: “Loop + Function” Input Validation

```python
# We reuse a function to keep the loop clean.
def input_marks():
    while True:
        s = input("Enter marks (0-100): ").strip()
        if s.isdigit() and 0 <= int(s) <= 100:
            return int(s)
        print("Invalid marks. Try again.")

m = input_marks()
print("Saved marks:", m)
```

## Part G — Docstrings

```python
# Docstrings describe what a function does.
def area_rectangle(length, width):
    """
    Returns the area of a rectangle.
    length: positive number
    width: positive number
    """
    return length * width

print(area_rectangle(5, 3))
```
