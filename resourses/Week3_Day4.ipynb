## Part A — Comprehension Basics

Comprehension is a short way to build a new list/set/dict using a loop.

```python
# Normal Loop vs Comprehension

nums = [1, 2, 3, 4, 5]

# loop way
squares1 = []
for n in nums:
    squares1.append(n * n)

# comprehension way
squares2 = [n * n for n in nums]

print(squares1)
print(squares2)
```

## Part B — List Comprehensions

### Example B1: Filter (keep only some)

```python
nums = [2, 5, 8, 11, 14, 17]
evens = [n for n in nums if n % 2 == 0]
print(evens)  # [2, 8, 14]
```

### Example B2: Transform (change items)

```python
names = ["fraz", "ALI", "zaIN"]
clean = [n.strip().title() for n in names]
print(clean)  # ['Fraz', 'Ali', 'Zain']
```

### Example B3: Filter + transform together

```python
nums = [1, 2, 3, 4, 5, 6]
squares_of_even = [n*n for n in nums if n % 2 == 0]
print(squares_of_even)  # [4, 16, 36]
```

### Example B4: Inline if-else (label items)

```python
# This is NOT filtering. It changes each element.

nums = [1, 2, 3, 4, 5]
labels = ["even" if n % 2 == 0 else "odd" for n in nums]
print(labels)
```

### Example B5: Flatten nested list

```python
matrix = [[1, 2, 3], [4, 5, 6]]
flat = [x for row in matrix for x in row]
print(flat)  # [1,2,3,4,5,6]
```

## Part C — Set Comprehension (unique results)

```python
text = "Python is easy and Python is powerful"
unique = {w.lower() for w in text.split()}
print(unique)
```

## Part D — Dict Comprehension

### Example D1: Build dict of squares

```python
nums = [1, 2, 3, 4]
sq = {n: n*n for n in nums}
print(sq)  # {1:1, 2:4, 3:9, 4:16}
```

### Example D2: Filter dict by value

```python
marks = {"Maria": 92, "Ali": 78, "Zain": 88, "Sara": 95}
top = {name: m for name, m in marks.items() if m >= 90}
print(top)  # {'Maria': 92, 'Sara': 95}
```

### Example D3: Change values in dict (apply rule)

```python
prices = {"pen": 50, "book": 200, "pencil": 30}
new_prices = {k: int(v * 1.10) for k, v in prices.items()}
print(new_prices)
```

## Part E — zip() (Combine lists)

### Example E1: Make pairs from two lists

```python
names = ["Maria", "Ali", "Zain"]
marks = [92, 78, 88]

pairs = list(zip(names, marks))
print(pairs)  # [('Maria',92), ('Ali',78), ('Zain',88)]
```

### Example E2: Build dict from two lists (common)

```python
result = dict(zip(names, marks))
print(result)  # {'Maria':92, 'Ali':78, 'Zain':88}
```

### Example E3: Unzip

```python
pairs = [("Maria", 92), ("Ali", 78)]
n, m = zip(*pairs)
print(n)  # ('Maria','Ali')
print(m)  # (92,78)
```

## Part F — enumerate() (Index + Value cleanly)

```python
# With index
items = ["pen", "book", "pencil"]
for i, item in enumerate(items):
    print(i, item)
```

```python
# Start index from 1
for i, item in enumerate(items, start=1):
    print(i, item)
```

```python
# Update list in-place using enumerate
prices = [100, 200, 300]
for i, p in enumerate(prices):
    prices[i] = int(p * 1.10)

print(prices)
```
