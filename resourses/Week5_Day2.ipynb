## Part A — Encapsulation

**What is encapsulation?**<br>
Encapsulation means:
- keeping data + related functions together inside a class
- controlling how data is accessed or changed

Simple idea: <br>
Instead of writing random variables everywhere, we keep them inside an object and control them through methods.

Example:
A bank account should not allow anyone to directly set balance to a wrong value like -5000.

## Part B — Public, Protected-style, and Private-style Members

### Example B1: Public attributes

```python
# These can be accessed directly from outside the class
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

s = Student("Maria", 92)
print(s.name)
print(s.marks)
```

```python
s.marks = -50 # Anyone can change values without control
print(s.marks)
```

### Example B2: Protected attributes (_name)

```python
# It is only a convention in Python. It can still be accessed.
class Student:
    def __init__(self, name, marks):
        self._name = name
        self._marks = marks

s = Student("Ali", 88)
print(s._name)
# Not truly private, used to show internal use
```

## Example B3: Private attributes (__name)

```python
# Double underscore triggers name mangling.
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

acc = BankAccount("Maria", 1000)
print(acc.owner)
# print(acc.__balance)   # Error
```

## Part C — Getters and Setters (Control access properly)

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = 0
        self.set_marks(marks)

    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Invalid marks. Setting to 0.")
            self.__marks = 0

    def get_marks(self):
        return self.__marks
```

```python
s = Student("Zain", 95)
print(s.get_marks())

s.set_marks(120)
print(s.get_marks())
```

## Part D — __ str__() (Neat object printing)

### Example D1: Without __ str__()

```python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

b = Book("Python Basics", "A. Khan", 850)
print(b)
```

### Example D2: With __ str__()

```python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    def __str__(self):
        return f"{self.title} by {self.author} - Rs. {self.price}"

b = Book("Python Basics", "A. Khan", 850)
print(b)
```

## Part E — Inheritance (One class reuses another)

**What is inheritance?**<br>
Inheritance means: one class takes features from another class

Example:
- Person is a general class
- Student is a special kind of Person

### Example E1: Basic inheritance

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Student(Person):
    def __init__(self, name, age, roll):
        super().__init__(name, age)
        self.roll = roll

    def student_info(self):
        print(f"Roll No: {self.roll}")

s = Student("Bilal", 22, 101)
s.show()
s.student_info()

# super().__init__(...) calls the parent class constructor.
```

### Example E2: Method overriding

```python
# A child class can change a method from parent. (Same method name, different behavior)

class Person:
    def intro(self):
        print("I am a person.")

class Teacher(Person):
    def intro(self):
        print("I am a teacher.")

p = Person()
t = Teacher()

p.intro()
t.intro()
```

## Part F — Composition vs Inheritance

**Composition means:**<br>
One class contains another object.

**Easy rule:**
- Inheritance = “is-a” relationship
- Composition = “has-a” relationship

Examples:
- Student is a Person → inheritance
- Car has an Engine → composition

```python
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

    def start(self):
        print(f"Engine with {self.horsepower} HP started.")

class Car:
    def __init__(self, brand, horsepower):
        self.brand = brand
        self.engine = Engine(horsepower)   # composition

    def start_car(self):
        print(f"{self.brand} car is starting...")
        self.engine.start()

c = Car("Toyota", 150)
c.start_car()

# It is useful because a car is not an engine. A car has an engine.
```

**Use inheritance when:**
- there is a real is-a relationship
- child truly extends parent behavior

Example:
- Teacher is a Person
- Dog is an Animal

**Use composition when:**
- one object just uses/contains another
- you want more flexible design

Example:
- Student has an Address
- Laptop has a Battery

## Part G — Intro to Operator Overloading

It gives special meaning to operators like +, -, == for your own objects.

## Example G1: Add two objects using __ add__()

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(2, 3)
p2 = Point(4, 5)

p3 = p1 + p2
print(p3)
```

## Example G2: Equality check using __ eq__()

```python
class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

    def __eq__(self, other):
        return self.roll == other.roll

s1 = Student("Maria", 101)
s2 = Student("Ali", 101)

print(s1 == s2)
# Because we decided that same roll number means same student.
```

## Common Mistakes

1. Thinking _name is truly private — it is not, it is only a convention
2. Using __private everywhere without understanding why
3. Using inheritance where composition is a better fit
4. Forgetting super().__ init__() in child class
5. Writing a bad __ str__() that does not return a string
6. Using operator overloading when normal methods would be simpler
