## Part A — What is a Data App?

A **data app** is a simple application that shows data in an interactive way.

Instead of only printing results in terminal or notebook, a data app lets users:

## Part E — First Streamlit App

Now we write our first Streamlit app file.

This app will show:
- title
- short text
- a number input
- a button
- a result

````python

from pathlib import Path

first_app_code = '''
import streamlit as st

st.title("My First Data App")
st.write("Welcome to Streamlit.")

num1 = st.number_input("Enter first number", value=0)
num2 = st.number_input("Enter second number", value=0)

if st.button("Add"):
    st.success(f"Result: {num1 + num2}")
'''

Path("app_first.py").write_text(first_app_code, encoding="utf-8")
print("Created file: app_first.py")
````

## Part H — Simple Sales Dashboard App

This app will show:
- table
- city filter
- summary numbers
- bar chart

````python

dashboard_app_code = '''
import streamlit as st
import pandas as pd

st.title("Sales Dashboard")

df = pd.DataFrame({
    "product": ["Shirt", "Shirt", "Shoes", "Bag", "Bag", "Shoes", "Shirt", "Watch"],
    "city": ["Lahore", "Karachi", "Lahore", "Islamabad", "Karachi", "Karachi", "Faisalabad", "Lahore"],
    "month": ["Jan", "Jan", "Jan", "Jan", "Feb", "Feb", "Feb", "Feb"],
    "sales": [12000, 15000, 18000, 9000, 11000, 20000, 14000, 25000]
})

selected_city = st.selectbox("Choose a city", ["All"] + sorted(df["city"].unique().tolist()))

if selected_city != "All":
    filtered_df = df[df["city"] == selected_city]
else:
    filtered_df = df.copy()

st.subheader("Filtered Data")
st.dataframe(filtered_df)

st.metric("Total Sales", int(filtered_df["sales"].sum()))
st.metric("Number of Records", len(filtered_df))

chart_data = filtered_df.groupby("product")["sales"].sum()
st.subheader("Sales by Product")
st.bar_chart(chart_data)
'''

Path("app_dashboard.py").write_text(dashboard_app_code, encoding="utf-8")
print("Created file: app_dashboard.py")
````

````python

print(Path("app_dashboard.py").read_text(encoding="utf-8"))
````

## Part J — Streamlit Forms

A **form** lets the user fill multiple inputs and then submit once.

This is useful when:
- we do not want app to update after every small change
- we want cleaner input handling
- we want a proper submission flow

````python

form_app_code = '''
import streamlit as st

st.title("Student Result Form")

with st.form("student_form"):
    name = st.text_input("Enter student name")
    marks = st.number_input("Enter marks", min_value=0, max_value=100, value=50)
    attendance = st.slider("Attendance %", 0, 100, 75)
    submitted = st.form_submit_button("Submit")

if submitted:
    if marks >= 50 and attendance >= 75:
        status = "Pass"
    else:
        status = "At Risk"

    st.write("Name:", name)
    st.write("Marks:", marks)
    st.write("Attendance:", attendance)
    st.success(f"Status: {status}")
'''

Path("app_form.py").write_text(form_app_code, encoding="utf-8")
print("Created file: app_form.py")
````

````python

print(Path("app_form.py").read_text(encoding="utf-8"))
````

## Part K — File Upload in Streamlit

One of the most useful features is **file upload**.

Students can make apps where user uploads:
- CSV file
- Excel file
- image
- text file

Then the app can:
- show preview
- clean data
- plot data
- calculate summaries

````python

upload_app_code = '''
import streamlit as st
import pandas as pd

st.title("CSV File Analyzer")

uploaded_file = st.file_uploader("Upload a CSV file", type=["csv"])

if uploaded_file is not None:
    df = pd.read_csv(uploaded_file)

    st.subheader("Preview")
    st.dataframe(df.head())

    st.subheader("Shape")
    st.write(df.shape)

    st.subheader("Columns")
    st.write(df.columns.tolist())

    numeric_cols = df.select_dtypes(include="number").columns.tolist()

    if numeric_cols:
        selected_col = st.selectbox("Choose numeric column for chart", numeric_cols)
        st.bar_chart(df[selected_col].head(10))
    else:
        st.warning("No numeric columns found.")
'''

Path("app_upload.py").write_text(upload_app_code, encoding="utf-8")
print("Created file: app_upload.py")
````

````python

print(Path("app_upload.py").read_text(encoding="utf-8"))
````

## Part N — Flask Templates Alternative

Not every app is made in Streamlit.

Another way is:
- use **Flask**
- create routes
- return HTML pages using templates

This gives more control, but usually needs more web knowledge than Streamlit.

````python

flask_app_code = '''
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/")
def home():
    students = [
        {"name": "Ali", "marks": 78},
        {"name": "Sara", "marks": 92},
        {"name": "Hina", "marks": 81}
    ]
    return render_template("index.html", students=students)

if __name__ == "__main__":
    app.run(debug=True)
'''
Path("flask_app.py").write_text(flask_app_code, encoding="utf-8")

templates_dir = Path("templates")
templates_dir.mkdir(exist_ok=True)

html_code = '''
<!DOCTYPE html>
<html>
<head>
    <title>Student Page</title>
</head>
<body>
    <h1>Student Marks</h1>
    <ul>
        {% for s in students %}
            <li>{{ s.name }} - {{ s.marks }}</li>
        {% endfor %}
    </ul>
</body>
</html>
'''
(templates_dir / "index.html").write_text(html_code, encoding="utf-8")

print("Created flask_app.py and templates/index.html")
````

````python

print(Path("flask_app.py").read_text(encoding="utf-8"))
print("\n--- index.html ---\n")
print((Path("templates") / "index.html").read_text(encoding="utf-8"))
````

## Part O — Streamlit vs Flask Templates

**Use Streamlit when:**
- you want quick dashboards
- you want to show data or ML results
- you want less frontend work
- you are making a student project or internal tool

**Use Flask templates when:**
- you want route control
- you want custom HTML pages
- you want more web-style app structure
- you may later grow into a bigger backend app

## Part P — Environment Variables

An **environment variable** is a value stored outside the code.

Examples:
- API key
- secret key
- database URL
- app mode (dev / prod)

Why use them?
- safer than hard-coding secrets
- easier to change values for different environments
- useful when deploying apps

````python

import os

os.environ["APP_NAME"] = "My Student Dashboard"
os.environ["APP_MODE"] = "development"

print("APP_NAME:", os.getenv("APP_NAME"))
print("APP_MODE:", os.getenv("APP_MODE"))
print("MISSING_VALUE:", os.getenv("MISSING_VALUE", "default_value"))
````

## Part Q — Why Hard-Coding Secrets Is Bad

Bad example:

```python
API_KEY = "12345SECRET"
```

Problems:
- secret becomes visible in code
- secret may be pushed to GitHub
- changing it later is harder

Better idea:

```python
import os
API_KEY = os.getenv("API_KEY")
```

## Part R — Using a `.env` File

A `.env` file is a simple text file that stores variables.

Example content:

```env
API_KEY=abc123
DB_URL=sqlite:///school.db
DEBUG=True
```

Then Python can read it using `python-dotenv`.

````python

env_text = '''
API_KEY=abc123_demo_key
DB_URL=sqlite:///school.db
DEBUG=True
'''.strip()

Path(".env").write_text(env_text, encoding="utf-8")
print("Created .env file")
print(Path(".env").read_text(encoding="utf-8"))
````

## Part T — Basic Deployment Idea

**Deployment** means putting the app on a server so other people can open it through a link.

Simple flow:
1. build the app on your computer
2. make sure files are organized
3. create `requirements.txt`
4. push code to GitHub
5. connect GitHub repo to a hosting platform
6. deploy and test

For beginners, the main idea is more important than platform details.

## Part U — Important Deployment Files

Usually projects need:
- main app file
