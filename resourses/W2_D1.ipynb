### Example C2: Parse key=value pairs (mini config / query parser)<a href="#Example-C2:-Parse-key=value-pairs-(mini-config-/-query-parser)" class="anchor-link">¶</a>

In \[ \]:

    raw = "name=Fraz; age=23; city=Lahore; role=Student"
    pairs = [p.strip() for p in raw.split(";") if p.strip()]

    data = {}
    for pair in pairs:
        key, value = pair.split("=", 1)   # split only once
        data[key.strip()] = value.strip()

    print(data)
    # {'name': 'Fraz', 'age': '23', 'city': 'Lahore', 'role': 'Student'}

### Example E1: Word frequency (basic text analytics)<a href="#Example-E1:-Word-frequency-(basic-text-analytics)" class="anchor-link">¶</a>

In \[ \]:

    import string

    text = "Python is great, and Python is fast! FAST fast."
    t = str.maketrans("", "", string.punctuation)

    clean = text.translate(t).casefold()
    words = clean.split()

    freq = {}
    for w in words:
        freq[w] = freq.get(w, 0) + 1

    print(freq)
    # {'python': 2, 'is': 2, 'great': 1, 'and': 1, 'fast': 3}
