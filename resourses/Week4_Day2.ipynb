## Part A — Parameters vs Arguments

```python
# Parameter: variable inside function definition
# Argument: actual value passed when calling function

def greet(name):      # name is parameter
    return f"Hello {name}"

print(greet("Maria")) # "Maria" is argument
```

## Part B — Positional vs Keyword Arguments

### Example B1: Positional arguments (order matters)

```python
def rectangle_area(length, width):
    return length * width

print(rectangle_area(5, 3))   # length=5, width=3
```

### Example B2: Keyword arguments (order does NOT matter)

```python
print(rectangle_area(width=3, length=5))  # same result
```

```python
# COMMON MISTAKE
rectangle_area(length=5, 3)  # ❌ Error: positional cannot come after keyword
# Rule: Positional arguments must come before keyword arguments.
```

## Part C — Default Parameters

### Example C1: Default Value

```python
def power(base, exp=2):
    return base ** exp

print(power(5))      # exp=2
print(power(5, 3))   # exp=3
print(power(base=2, exp=5))
```

### Example C2: Multiple defaults (designing a clean API)

```python
def format_name(first, last, middle=""):
    if middle:
        return f"{first.title()} {middle.title()} {last.title()}"
    return f"{first.title()} {last.title()}"

print(format_name("maria", "nazir"))
print(format_name("maria", "nazir", middle="fatima"))
```

### Important Pitfall: Mutable Default Parameter (BUG)

```python
def add_item(item, basket=[]):
    basket.append(item)
    return basket

print(add_item("pen"))
print(add_item("book"))  # surprising result!
```

Why this happens: default list is created once and reused.

```python
# Correct Version
def add_item(item, basket=None):
    if basket is None:
        basket = []
    basket.append(item)
    return basket

print(add_item("pen"))
print(add_item("book"))
```

## Part D — *args (Variable Positional Arguments)

### Example D1: Sum any number of values

```python
def total(*args):
    s = 0
    for x in args:
        s += x
    return s

print(total(1, 2, 3))
print(total(10, 20))
print(total())  # works, returns 0
# args becomes a tuple inside the function.
```

### Example D2: Mix normal parameters + *args

```python
def average(title, *nums):
    if not nums:
        return f"{title}: No numbers"
    return f"{title}: {sum(nums) / len(nums):.2f}"

print(average("Test1", 10, 20, 30))
print(average("Test2"))
```

### Example D3: Unpacking a list into *args

```python
nums = [2, 4, 6, 8]
print(total(*nums))  # same as total(2,4,6,8)
```

## Part E — **kwargs (Variable Keyword Arguments)

### Example E1: Print a profile from flexible data

```python
def profile(**kwargs):
    for k, v in kwargs.items():
        print(f"{k:<10}: {v}")

profile(name="Maria", age=22, city="Lahore")
# kwargs becomes a dictionary.
```

### Example E2: Default values using kwargs + get()

```python
def student_report(**info):
    name = info.get("name", "Unknown")
    marks = info.get("marks", 0)
    city = info.get("city", "N/A")
    print(f"Name : {name}")
    print(f"Marks: {marks}")
    print(f"City : {city}")

student_report(name="Ali", marks=78)
student_report(city="Peshawar", marks=88, name="Zain")
```

### Example E3: Unpacking a dictionary into **kwargs

```python
data = {"name": "Sara", "marks": 95, "city": "Islamabad"}
student_report(**data)
```
