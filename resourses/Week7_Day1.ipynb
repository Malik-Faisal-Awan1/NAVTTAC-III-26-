## Part A — Why Pandas Matters

When data becomes bigger than a few variables, plain Python becomes slow and messy for analysis.

Pandas helps us work with:

- student records
- shop sales
- employee data
- survey responses
- sensor readings
- CSV and Excel files

In simple words:

- a **Series** is like one column of data
- a **DataFrame** is like a full table

So Pandas is very useful in data analysis, machine learning, business reports, and real projects.


```python
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
print("NumPy version:", np.__version__)
```

## Part B — First Understand the Two Main Structures

### 1) Series
A **Series** is a one-dimensional labeled data structure.

You can think of it like:

- one column of a table
- or a list with an index

### 2) DataFrame
A **DataFrame** is a two-dimensional labeled table.

You can think of it like:

- rows and columns in Excel
- a full dataset


## Part C — Create a Simple Series

Let us start with the easiest possible example.


```python
marks = pd.Series([78, 85, 92, 88, 90])
print(marks)
```

### What should students notice?

- Values are shown in one column
- The numbers on the left are the **index**
- The numbers on the right are the actual **data values**


## Part D — Check Properties of a Series
A Series also has useful properties and methods.


```python
marks = pd.Series([78, 85, 92, 88, 90])

print("Series:")
print(marks)

print("\nType:", type(marks))
print("Shape:", marks.shape)
print("Size:", marks.size)
print("Data type:", marks.dtype)
```

## Part E — Create a Series with Custom Index

Instead of default index `0, 1, 2, ...`, we can give meaningful labels.


```python
marks = pd.Series([78, 85, 92], index=["Ali", "Sara", "Ahmed"])
print(marks)
```

## Part F — Access Values from a Series

We can access values by label or by position.


```python
marks = pd.Series([78, 85, 92], index=["Ali", "Sara", "Ahmed"])

print("Ali's marks:", marks["Ali"])
print("First value:", marks.iloc[0])
```

### Important note

For beginners, it is good to remember:

- label-based access -> use names like `"Ali"`
- position-based access -> use `.iloc[]`


## Part G — Create a Series from a Dictionary

This is another very common way.


```python
data = {
    "Math": 90,
    "English": 84,
    "Physics": 88
}

subject_marks = pd.Series(data)
print(subject_marks)
```

## Part H — Series Summary Methods

A Series can quickly give useful information.


```python
marks = pd.Series([78, 85, 92, 88, 90])

print("Sum:", marks.sum())
print("Mean:", marks.mean())
print("Minimum:", marks.min())
print("Maximum:", marks.max())
print("Standard deviation:", marks.std())
```

## Part I — Now Move to DataFrame

A DataFrame is like a full table with many columns.


```python
students = {
    "Name": ["Ali", "Sara", "Ahmed", "Hina", "Usman"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [78, 85, 92, 88, 90],
    "City": ["Lahore", "Karachi", "Lahore", "Islamabad", "Karachi"]
}

df = pd.DataFrame(students)
print(df)
```

## Part J — Understand the Table Structure

This table has:

- rows -> each student record
- columns -> `Name`, `Age`, `Marks`, `City`
- index -> row labels on the left


## Part K — Basic DataFrame Properties

These are the first things students should check after loading a dataset.


```python
print("Shape:", df.shape)
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
print("\nColumns:")
print(df.columns)
print("\nIndex:")
print(df.index)
print("\nData types:")
print(df.dtypes)
```

## Part L — Look at the First and Last Rows

This is very useful when the dataset is large.


```python
print("First 3 rows:")
print(df.head(3))

print("\nLast 2 rows:")
print(df.tail(2))
```

## Part M — Select One Column

Selecting one column returns a **Series**.


```python
names = df["Name"]
print(names)
print("\nType:", type(names))
```

## Part N — Select Multiple Columns

Selecting multiple columns returns a **DataFrame**.


```python
selected = df[["Name", "Marks"]]
print(selected)
print("\nType:", type(selected))
```

## Part O — Add a New Column

We can easily create a new column from old columns or fixed values.


```python
df["Pass"] = df["Marks"] >= 80
print(df)
```

## Part P — Add Another Calculated Column

Let us create a grade-style column using `np.where()`.


```python
df["Result"] = np.where(df["Marks"] >= 80, "Pass", "Fail")
print(df)
```

## Part Q — Select Rows by Position with `.iloc[]`

`.iloc[]` means **integer location**.

It works with row and column positions.


```python
print("First row:")
print(df.iloc[0])

print("\nFirst three rows:")
print(df.iloc[0:3])

print("\nRows 1 to 3 and columns 0 to 2:")
print(df.iloc[1:4, 0:3])
```

## Part R — Select Rows by Label with `.loc[]`

`.loc[]` works with labels.

Right now the row labels are `0, 1, 2, 3, 4`, so label and position look similar.

Later, if we change the index, `.loc[]` becomes even more meaningful.


```python
print("Row with label 2:")
print(df.loc[2])

print("\nRows with labels 1 to 3:")
print(df.loc[1:3])

print("\nRows 1 to 3 and selected columns:")
print(df.loc[1:3, ["Name", "Marks", "Result"]])
```

## Part S — Set a Meaningful Index

Sometimes a column should act like the row label.


```python
df_by_name = df.set_index("Name")
print(df_by_name)
```

## Part T — Use `.loc[]` After Setting the Index

Now row labels are student names.


```python
print(df_by_name.loc["Sara"])
print("\nOnly selected columns for Ahmed:")
print(df_by_name.loc["Ahmed", ["Age", "Marks", "City"]])
```

### Important idea

Remember this difference clearly:

- `.loc[]` -> works with labels
- `.iloc[]` -> works with number positions

This is one of the most important Pandas concepts.


## Part U — Quick Filtering with Conditions

This topic becomes bigger later, but students should at least see the basic idea today.


```python
high_marks = df[df["Marks"] > 85]
print(high_marks)
```

## Part V — Multiple Conditions

Use `&` for AND and `|` for OR.

Also remember to put each condition inside brackets.


```python
filtered = df[(df["Marks"] > 80) & (df["City"] == "Karachi")]
print(filtered)
```

## Part W — Sort Values

Sorting helps us understand highest, lowest, and ordered records.


```python
print("Sort by marks ascending:")
print(df.sort_values("Marks"))

print("\nSort by marks descending:")
print(df.sort_values("Marks", ascending=False))
```

## Part X — Rename Columns

This is useful when column names are unclear.


```python
df_renamed = df.rename(columns={"Marks": "Final_Marks"})
print(df_renamed.columns)
print(df_renamed)
```

## Part Y — Read Data from a CSV File

In real work, data usually comes from files.

For this lab, we will first create a small sample CSV file, then read it using Pandas.


```python
sample_sales = pd.DataFrame({
    "Product": ["Pen", "Book", "Bag", "Bottle", "Notebook"],
    "Price": [20, 120, 1500, 350, 80],
    "Quantity": [10, 5, 2, 6, 12],
    "City": ["Lahore", "Karachi", "Lahore", "Islamabad", "Karachi"]
})

sample_sales.to_csv("sample_sales.csv", index=False)
print("CSV file created successfully")
```

```python
sales_csv = pd.read_csv("sample_sales.csv")
print(sales_csv)
```

## Part Z — Understand Common `read_csv()` Usage

Very common patterns are:

- `pd.read_csv("file.csv")`
- `pd.read_csv("file.csv", usecols=[...])`
- `pd.read_csv("file.csv", nrows=...)`

Let us try some simple ones.


```python
print("Only selected columns:")
print(pd.read_csv("sample_sales.csv", usecols=["Product", "Price"]))

print("\nRead first 3 rows only:")
print(pd.read_csv("sample_sales.csv", nrows=3))
```

## Part AA — Read Data from an Excel File

Pandas can also read Excel files.

We will create a sample Excel file first.


```python
sample_sales.to_excel("sample_sales.xlsx", index=False)
print("Excel file created successfully")
```

```python
sales_excel = pd.read_excel("sample_sales.xlsx")
print(sales_excel)
```

### Note for students

If Excel reading ever gives an engine-related error on some system, it usually means a required package is missing.

But in most normal lab setups, this works fine.


## Part AB — Compare CSV and Excel Reading

The syntax is very similar:

- CSV -> `pd.read_csv()`
- Excel -> `pd.read_excel()`

That is why Pandas is very convenient.


## Part AC — Data Types in a DataFrame

Data types matter because they affect calculations and analysis.

Common types include:

- `int64` for whole numbers
- `float64` for decimal numbers
- `object` for text
- `bool` for True/False


```python
print(df.dtypes)
```

## Part AD — Convert a Data Type with `astype()`

Sometimes a column needs conversion.

For example, maybe a numeric column is stored as text.


```python
temp_df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Marks": ["78", "85", "90"]
})

print("Before conversion:")
print(temp_df.dtypes)

temp_df["Marks"] = temp_df["Marks"].astype(int)

print("\nAfter conversion:")
print(temp_df.dtypes)
print(temp_df)
```

## Part AE — Basic Dataset Summary with `info()`

`info()` is one of the most useful commands in Pandas.

It shows:

- number of rows
- number of columns
- column names
- non-null counts
- data types


```python
df.info()
```

## Part AF — Basic Numerical Summary with `describe()`

`describe()` gives quick statistics for numeric columns.


```python
print(df.describe())
```

## Part AG — Include Non-Numeric Columns Too

If you want a broader summary, use `include="all"`.


```python
print(df.describe(include="all"))
```

## Part AH — Other Useful Quick Summary Methods

These methods are simple but very practical.


```python
print("Average marks:", df["Marks"].mean())
print("Highest marks:", df["Marks"].max())
print("Lowest marks:", df["Marks"].min())
print("Total students:", df["Name"].count())
print("Unique cities:", df["City"].nunique())
```

## Part AI — Frequency Count of a Column

`value_counts()` is very useful for categorical data.


```python
print(df["City"].value_counts())
```

## Part AJ — Unique Values in a Column

Sometimes we want to see all possible categories.


```python
print("Unique cities:", df["City"].unique())
print("Number of unique cities:", df["City"].nunique())
```

## Part AN — Common Beginner Mistakes

Students often make these mistakes:

1. Forgetting the double brackets when selecting multiple columns  
   Wrong: `df["Name", "Marks"]`  
   Correct: `df[["Name", "Marks"]]`

2. Mixing `.loc[]` and `.iloc[]`  
   `.loc[]` is for labels  
   `.iloc[]` is for positions

3. Expecting one column selection to return a DataFrame  
   `df["Name"]` gives a Series  
   `df[["Name"]]` gives a DataFrame

4. Forgetting brackets around each condition in filtering

5. Ignoring `dtypes` and doing calculations on text columns
