## Part A — Basics + Mutability

### Example A1: Creating lists

```python
nums = [10, 20, 30]
mixed = ["Fraz", 22, True, 3.5]
empty = []
from_range = list(range(1, 6))

print(nums)
print(mixed)
print(from_range)
```

### Example A2: Indexing and slicing

```python
arr = ["a", "b", "c", "d", "e"]

print(arr[0])     # a
print(arr[-1])    # e
print(arr[1:4])   # ['b', 'c', 'd']
print(arr[:3])    # first 3
print(arr[::2])   # every 2nd
print(arr[::-1])  # reversed copy
```

### Example A3: Lists are mutable

```python
names = ["Ali", "Zain", "Maria"]
names[1] = "Haseeb"     # allowed
print(names)
```

## Part B — List Methods (core toolkit)

### Example B1: append vs extend

```python
a = [1, 2, 3]
a.append(4)        # adds ONE item
print(a)           # [1,2,3,4]

b = [1, 2, 3]
b.extend([4, 5])   # adds multiple items
print(b)           # [1,2,3,4,5]
```

### Example B2: insert

```python
x = [10, 20, 30]
x.insert(1, 99)  # insert at index 1
print(x)         # [10, 99, 20, 30]
```

### Example B3: pop and remove

```python
data = ["a", "b", "c", "b"]

data.pop()        # removes last
print(data)

data.pop(1)       # removes at index 1
print(data)

data.remove("b")  # removes first occurrence of value "b"
print(data)
```

### Example B4: clear

```python
lst = [1, 2, 3]
lst.clear()
print(lst)  # []
```

### Example B5: index and count

```python
lst = ["x", "y", "x", "z"]
print(lst.count("x"))     # 2
print(lst.index("y"))     # 1
```

### Example B6: sort vs sorted

```python
nums = [5, 2, 9, 1]
nums.sort()          # changes original
print(nums)

nums2 = [5, 2, 9, 1]
new_nums = sorted(nums2)  # returns new list
print(nums2)              # unchanged
print(new_nums)
```

### Example B7: sort with key

```python
names = ["maria", "Ali", "zain", "Haseeb"]
print(sorted(names))                 # default (ASCII)
print(sorted(names, key=str.lower))  # case-insensitive
```

## Part C — Copying Lists

### Example C1: Assignment creates shared reference (bug!)

```python
a = [1, 2, 3]
b = a           # NOT a copy
b.append(99)

print(a)  # [1,2,3,99]  <-- changed!
print(b)
```

### Example C2: Correct copying

```python
a = [1, 2, 3]

b = a.copy()   # or list(a) or a[:]
b.append(99)

print(a)  # [1,2,3]
print(b)  # [1,2,3,99]
```

## Part D — Nested Lists

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

print(matrix[0])     # [1,2,3]
print(matrix[0][1])  # 2
```

## Part E — Real Patterns with Lists (Practical Skills)

### Example E1: Searching in a list

```python
items = ["pen", "book", "pencil"]
target = input("Enter item: ").strip().lower()

found = False
for x in items:
    if x == target:
        found = True
        break

print("Found!" if found else "Not found.")
```

### Example E2: Filter items (build a new list)

```python
nums = [2, 5, 8, 11, 14, 17]
evens = []

for n in nums:
    if n % 2 == 0:
        evens.append(n)

print(evens)
```

### Example E3: Update items in a list (in-place)

```python
prices = [100, 200, 300]
for i in range(len(prices)):
    prices[i] = prices[i] * 1.10  # add 10%

print(prices)
```
