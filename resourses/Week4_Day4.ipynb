## Part A — Functions are “Values” in Python

### Example A1: Pass a function into another function

```python
def apply_twice(func, x):
    return func(func(x))

def add_one(n):
    return n + 1

print(apply_twice(add_one, 5))  # 7
```

### Example A2: Same thing using lambda (short)

```python
print(apply_twice(lambda n: n + 1, 5))  # 7
```

```python
# lambda is just a short anonymous function:
lambda n: n + 1
# Means: “a function that takes n and returns n + 1”
```

## Part B — map() (Transform each element)

map(function, iterable) applies the function to every item.

### Example B1: Square every number (normal function)

```python
def square(n):
    return n * n

nums = [1, 2, 3, 4]
result = list(map(square, nums))
print(result)  # [1, 4, 9, 16]
```

### Example B2: Square every number (lambda)

```python
nums = [1, 2, 3, 4]
result = list(map(lambda n: n * n, nums))
print(result)  # [1, 4, 9, 16]
```

### Example B3: Clean and title-case names using map + lambda

```python
raw = ["  maria  ", "ALI", "  zain  "]
cleaned = list(map(lambda s: " ".join(s.strip().split()).title(), raw))
print(cleaned)  # ['Maria', 'Ali', 'Zain']
```

### Example B4: map() with multiple iterables

```python
a = [1, 2, 3]
b = [10, 20, 30]
sum_list = list(map(lambda x, y: x + y, a, b))
print(sum_list)  # [11, 22, 33]
# map() stops at the shortest list if lengths are different.
```

## Part C — filter() (Keep only some elements)

filter(condition_function, iterable) keeps items where the function returns True.

### Example C1: Keep only even numbers (lambda)

```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda n: n % 2 == 0, nums))
print(evens)  # [2, 4, 6]
```

### Example C2: Keep only valid marks from strings (0–100)

```python
raw_marks = ["92", "abc", "120", "0", "78", "-5", " 55 "]

# Clean spaces first
cleaned = list(map(lambda s: s.strip(), raw_marks))

# Keep only digit strings AND range 0..100
valid = list(filter(lambda s: s.isdigit() and 0 <= int(s) <= 100, cleaned))
print(valid)  # ['92', '0', '78', '55']

marks = list(map(int, valid))
print(marks)  # [92, 0, 78, 55]
```

### Example C3: Keep non-empty cleaned strings

```python
raw = ["  hello  ", "   ", "python", "", "  AI  "]

cleaned = list(map(lambda s: " ".join(s.strip().split()), raw))
non_empty = list(filter(lambda s: s != "", cleaned))

print(cleaned)
print(non_empty)
```

## Part D — sorted() with key= (Sorting the correct way)

Sorting basics
- sorted(x) returns a new list
- x.sort() changes the original list
- key= tells Python what to sort by

### Example D1: Case-insensitive sort (lambda)

```python
names = ["maria", "Ali", "zain", "Haseeb"]
print(sorted(names, key=lambda s: s.lower()))
```

### Example D2: Sort list of tuples by marks (second item)

```python
students = [("Maria", 92), ("Ali", 78), ("Zain", 88)]
by_marks = sorted(students, key=lambda t: t[1], reverse=True)
print(by_marks)
```

### Example D3: Sort dictionary items by value (frequency sort)

```python
freq = {"python": 2, "fast": 3, "and": 1, "is": 2}
items_sorted = sorted(freq.items(), key=lambda kv: kv[1], reverse=True)
print(items_sorted)  # [('fast', 3), ('python', 2), ('is', 2), ('and', 1)]
```

### Example D4: Multi-sort in one line (marks desc, name asc)

```python
students = [("Ali", 90), ("Sara", 95), ("Zain", 90), ("Ayesha", 90)]

# Sort by marks descending, then name ascending
result = sorted(students, key=lambda t: (-t[1], t[0]))
print(result)

# Trick: Use -t[1] to sort marks descending while still using sorted() default ascending.
```

## Part E — any() and all() (Clean checks)

### Example E1: Check if any student failed (< 50)

```python
marks = [78, 45, 92, 60]
has_fail = any(m < 50 for m in marks)
print(has_fail)
```

### Example E2: Check if all marks are valid (0–100)

```python
marks = [0, 55, 100, 88]
all_valid = all(0 <= m <= 100 for m in marks)
print(all_valid)
```

## Part F — Mini Pipeline Program (Clean → Filter → Convert → Report)

**Problem**

You get messy marks. Build a pipeline:
- clean spaces
- keep only valid marks
- convert to int
- sort and print average

```python
raw = [" 92 ", "abc", " 78", "  101 ", " 0 ", " 55 ", " -2 "]

cleaned = list(map(lambda s: s.strip(), raw))
valid = list(filter(lambda s: s.isdigit() and 0 <= int(s) <= 100, cleaned))
marks = list(map(int, valid))

marks_sorted = sorted(marks, reverse=True)

print("Cleaned:", cleaned)
print("Valid  :", valid)
print("Marks  :", marks_sorted)

if marks_sorted:
    avg = sum(marks_sorted) / len(marks_sorted)
    print(f"Average: {avg:.2f}")
else:
    print("No valid marks found.")
```

## When to use lambda (simple rule)

✅ Use lambda when:
- it’s short (1 line) and very clear
- you’re using it once inside map/filter/sorted

❌ Avoid lambda when:
- the logic is long/complex
- you need debugging or reusability
- In those cases, write a normal def.
