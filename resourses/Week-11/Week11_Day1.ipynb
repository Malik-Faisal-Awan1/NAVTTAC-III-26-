## Part A — What is the Web?

Very simple idea:

- when you open a site in browser, your browser asks for data
- a machine on the internet sends the data back
- that machine is called a **server**
- your browser or app is the **client**

So the web works because a **client** sends a **request** and a **server** sends a **response**.

## Part B — Website, Web App, and API

### Website
A website mostly shows information to users.

Example:
- college website
- news website
- clothing brand website

### Web App
A web app lets users do actions.

Example:
- Gmail
- Facebook

- online attendance system

### API
API stands for **Application Programming Interface**.

Simple meaning:
an API lets one software talk to another software.

Examples:
- mobile app sends login data to server
- weather app requests weather data from server
- student portal asks server for result details

## Part C — What is REST?

REST is a common way to design APIs.

You do not need hard theory here.

Just remember this simple idea:

- every useful thing in the system is treated like a **resource**
- each resource has a URL
- we use HTTP methods to perform actions on that resource

Examples of resources:
- students
- products
- orders
- books

Examples of URLs:
- `/students`
- `/products`
- `/orders/5`

So REST usually means:

- `GET /students` → read students
- `POST /students` → add a new student
- `PUT /students/5` → update student 5
- `DELETE /students/5` → delete student 5

### Response

## Part F — Common Status Codes

These numbers help us understand what happened.

- `200 OK` → request worked
- `201 Created` → new data created successfully
- `400 Bad Request` → client sent wrong or incomplete data
- `404 Not Found` → requested route or data not found
- `500 Internal Server Error` → server-side problem

Students should not memorize too many codes at once.
These 5 are enough for today.

````python
status_meanings = {
    200: "OK - request succeeded",
    201: "Created - new data added successfully",
    400: "Bad Request - wrong or incomplete client data",
    404: "Not Found - route or item not found",
    500: "Internal Server Error - problem on server side"
}

for code_num, meaning in status_meanings.items():
    print(f"{code_num}: {meaning}")
````

## Part I — Path Parameter vs Query Parameter

Example:

Here `5` is often an ID.

### Query parameter
It comes after `?` in URL.

Example:
- `/search?name=ali&city=lahore`

Here:
- `name=ali`
- `city=lahore`

Both are query parameters.

````python
path_example = "/students/5"
query_example = "/search?name=ali&city=lahore"

print("Path parameter example:", path_example)
print("Query parameter example:", query_example)
````

## Part K — Create a Very Small Flask App

We will create a mini API for student data.

We will not run a real internet server inside this notebook.

Instead, we will use **Flask test client**.

This is very useful in a notebook because:
- no separate browser tab is required
- no Postman is required for basic testing
- students can see request and response directly in notebook

````python
app = Flask(__name__)

students = [
    {"id": 1, "name": "Ali", "city": "Lahore", "age": 20},
    {"id": 2, "name": "Sara", "city": "Karachi", "age": 22},
    {"id": 3, "name": "Ahmed", "city": "Faisalabad", "age": 21}
]
````

## Part L — Route 1: Home Route

This route simply checks whether our API is alive.

It returns a welcome message.

````python
@app.route("/", methods=["GET"])
def home():
    return jsonify({
        "message": "Welcome to the student API",
        "status": "running"
    }), 200
````

## Part M — Route 2: Get All Students

This route returns the full student list.

Method used:
- `GET`

````python
@app.route("/students", methods=["GET"])
def get_students():
    return jsonify({
        "count": len(students),
        "data": students
    }), 200
````

## Part N — Route 3: Get One Student by ID

This uses a **path parameter**.

Example:
- `/students/2`

If the student exists, return the student.
If not, return `404`.

````python
@app.route("/students/<int:student_id>", methods=["GET"])
def get_student(student_id):
    for student in students:
        if student["id"] == student_id:
            return jsonify(student), 200

    return jsonify({
        "error": "Student not found"
    }), 404
````

## Part O — Route 4: Search Students by City

This uses a **query parameter**.

Example:
- `/search?city=Lahore`

If city is missing, we return `400`.

````python
@app.route("/search", methods=["GET"])
def search_by_city():
    city = request.args.get("city")

    if not city:
        return jsonify({
            "error": "city query parameter is required"
        }), 400

    result = [student for student in students if student["city"].lower() == city.lower()]

    return jsonify({
        "count": len(result),
        "data": result
    }), 200
````

## Part P — Route 5: Add a New Student

This route uses:
- `POST`
- JSON body

The client must send data like:

```json
{
  "name": "Hina",
  "city": "Multan",
  "age": 19
}
```

We will also do simple validation.

````python
@app.route("/students", methods=["POST"])
def add_student():
    data = request.get_json(silent=True)

    if not data:
        return jsonify({"error": "JSON body is required"}), 400

    required_fields = ["name", "city", "age"]
    missing = [field for field in required_fields if field not in data]

    if missing:
        return jsonify({
            "error": "Missing required fields",
            "missing_fields": missing
        }), 400

    new_id = max(student["id"] for student in students) + 1 if students else 1

    new_student = {
        "id": new_id,
        "name": data["name"],
        "city": data["city"],
        "age": data["age"]
    }

    students.append(new_student)

    return jsonify({
        "message": "Student created successfully",
        "student": new_student
    }), 201
````

## Part Q — Route 6: Update Student City

This gives students light exposure to `PUT`.

We will update the city of a student by ID.

````python
@app.route("/students/<int:student_id>", methods=["PUT"])
def update_student(student_id):
    data = request.get_json(silent=True)

    if not data or "city" not in data:
        return jsonify({"error": "city is required in JSON body"}), 400

    for student in students:
        if student["id"] == student_id:
            student["city"] = data["city"]
            return jsonify({
                "message": "Student updated successfully",
                "student": student
            }), 200

    return jsonify({"error": "Student not found"}), 404
````

## Part R — Route 7: Delete Student

This gives students light exposure to `DELETE`.

````python
@app.route("/students/<int:student_id>", methods=["DELETE"])
def delete_student(student_id):
    for i, student in enumerate(students):
        if student["id"] == student_id:
            removed = students.pop(i)
            return jsonify({
                "message": "Student deleted successfully",
                "student": removed
            }), 200

    return jsonify({"error": "Student not found"}), 404
````
