## Part A — Scope Basics (Local vs Global)

```python
# Local Scope (inside function)
def demo():
    x = 10  # local
    print("Inside:", x)

demo()
# print(x)  # Error: x not defined outside
```

```python
# Global Scope (outside function)
x = 99  # global

def show():
    print("Inside:", x)  # can read global

show()
print("Outside:", x)
```

## Part B — Modifying Global Variables

### Example B1: Without global (creates local variable)

```python
count = 0

def inc():
    count = count + 1  # ❌ Error: count referenced before assignment

# inc()
```

Why? Because Python thinks count is local (due to assignment), but it’s not set locally yet.

### Example B2: Correct using global

```python
count = 0

def inc():
    global count
    count = count + 1

inc()
inc()
print(count)  # 2

# global works, but avoid it for normal programs. Better: return a value and update outside.
```

## Part C — LEGB Rule (Where Python searches variables)

Python searches names in this order:
- L = Local
- E = Enclosing (outer function)
- G = Global
- B = Built-in

```python
x = "GLOBAL"

def outer():
    x = "ENCLOSING"
    def inner():
        x = "LOCAL"
        print(x)
    inner()

outer()  # prints LOCAL
```

## Part D — nonlocal (Enclosing scope modification)

### Example D1: Problem without nonlocal

```python
def outer():
    x = 10
    def inner():
        x = x + 1   # ❌ Error (same reason as global case)
    inner()

# outer()
```

### Example D2: Fix using nonlocal

```python
def outer():
    x = 10
    def inner():
        nonlocal x
        x = x + 1
        print("Inner x:", x)
    inner()
    print("Outer x:", x)

outer()

# nonlocal updates the variable in the enclosing function.
```

## Part E — Recursion Fundamentals

What is recursion? A function that calls itself.

Every recursion needs:
- Base case (stop condition)
- Recursive step (move toward base case)

### Example E1: Factorial (classic)

```python
def factorial(n):
    if n == 0 or n == 1:   # base case
        return 1
    return n * factorial(n - 1)  # recursive step

print(factorial(5))
```

### Example E2: Sum from 1 to n

```python
def sum_n(n):
    if n == 1:
        return 1
    return n + sum_n(n - 1)

print(sum_n(5))  # 15
```

### Example E3: Power function (base^exp)

```python
def power(base, exp):
    if exp == 0:
        return 1
    return base * power(base, exp - 1)

print(power(2, 5))  # 32
```

## Part F — Recursion with Lists and Strings

### Example F1: Sum a list recursively

```python
def sum_list(nums):
    if not nums:      # base case: empty list
        return 0
    return nums[0] + sum_list(nums[1:])

print(sum_list([2, 4, 6, 8]))
```

### Example F2: Reverse a string recursively

```python
def reverse_str(s):
    if s == "":
        return ""
    return reverse_str(s[1:]) + s[0]

print(reverse_str("python"))
```

### Example F3: Palindrome check recursively

```python
def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])

print(is_palindrome("madam"))
print(is_palindrome("python"))
```
