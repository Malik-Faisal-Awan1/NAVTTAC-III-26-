## Part A — Truthy / Falsey + Short-Circuit

### Example A1: Truthy/Falsey (important concept)

In Python:

0, "", None, [], {}, set() → False

Everything else → True

```python
print(bool(0))        # False
print(bool(""))       # False
print(bool([]))       # False
print(bool("Hi"))     # True
print(bool([1, 2]))   # True
```

```python
name = input("Enter name: ").strip()
if not name:
    print("Name is required!")
else:
    print("Welcome:", name)
```

### Example A2: Short-circuit logic (safe checking)

Short-circuit means:

In A and B, if A is False, Python does not check B

In A or B, if A is True, Python does not check B

```python
text = ""
# Safe: second part won't run because text is empty (False)
if text and text[0] == "A":
    print("Starts with A")
else:
    print("Empty or not starting with A")
```

## Part B — Clean Conditions Patterns (Real programs)

### Example B1: Membership checking (fast + clean)

```python
day = input("Enter day: ").strip().lower()
if day in ("saturday", "sunday"):
    print("Weekend")
else:
    print("Weekday")
```
