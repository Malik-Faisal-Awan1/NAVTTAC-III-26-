## Part A — OOP in Simple Words

- A class is a blueprint/template.
- An object is a real thing created from that blueprint.

Example:
- Class: Student
- Objects: Student("Maria", 101) and Student("Ali", 102)

## Part B — First Class (Attributes + Methods)

### Example B1: Create a simple class

```python
class Person:
    def __init__(self, name, age):
        self.name = name     # instance attribute
        self.age = age       # instance attribute

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")

p1 = Person("Umer", 22)
p2 = Person("Ali", 20)

p1.show()
p2.show()
```

What is self?
- self means “this object”
- It lets the object store and access its own data

### Example B2: Updating attributes

```python
p = Person("Zain", 19)
p.show()

p.age = 20
p.show()
```

## Part C — Common Mistake (Forgetting self)

```python
class Test:
    def __init__(self, x):
        self.x = x

    def show():
        print("Hello")   # ❌ wrong (missing self)
```

```python
# ✅ Correct:
class Test:
    def __init__(self, x):
        self.x = x

    def show(self):
        print("Hello")
```

## Part D — Instance Attributes vs Class Attributes

### Example D1: Class attributes (shared)

```python
class Student:
    school = "ABC College"   # class attribute (shared by all)

    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

s1 = Student("Umer", 101)
s2 = Student("Ali", 102)

print(s1.school)
print(s2.school)
print(Student.school)
```

```python
# Change class attribute
Student.school = "New College"
print(s1.school)
print(s2.school)
```

### Example D2: Instance attribute overrides class attribute

```python
s1.school = "Private School"
print(s1.school)      # instance version
print(s2.school)      # class version
print(Student.school) # class version

# Student.school is shared
# s1.school becomes separate only for that object if you assign it
```

## Part E — Real Examples

### Example E1: Student with methods

```python
class Student:
    def __init__(self, name, roll, marks):
        self.name = " ".join(name.strip().split()).title()
        self.roll = roll
        self.marks = marks if 0 <= marks <= 100 else 0

    def grade(self):
        if self.marks >= 90:
            return "A"
        if self.marks >= 80:
            return "B"
        if self.marks >= 70:
            return "C"
        return "D"

    def report(self):
        print("\n--- STUDENT REPORT ---")
        print(f"Name : {self.name}")
        print(f"Roll : {self.roll}")
        print(f"Marks: {self.marks}")
        print(f"Grade: {self.grade()}")

st = Student("  aLi   naZir ", 101, 92)
st.report()
```

### Example E2: BankAccount with deposit/withdraw

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = " ".join(owner.strip().split()).title()
        self.balance = balance if balance >= 0 else 0

    def deposit(self, amount):
        if amount <= 0:
            print("Deposit must be positive.")
            return
        self.balance += amount
        print(f"Deposited {amount}. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdraw must be positive.")
            return
        if amount > self.balance:
            print("Not enough balance.")
            return
        self.balance -= amount
        print(f"Withdrawn {amount}. New balance: {self.balance}")

    def show(self):
        print(f"Owner: {self.owner}, Balance: {self.balance}")
```

```python
acc = BankAccount(" ali nazir ", 1000)
acc.show()
acc.deposit(500)
acc.withdraw(300)
acc.withdraw(5000)
acc.show()
```
